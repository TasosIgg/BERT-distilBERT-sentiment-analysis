```python
!pip install scikit-learn
!pip install transformers
!pip install emoji contractions
!pip install datasets


import pandas as pd
import numpy as np
import torch
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_linear_schedule_with_warmup

from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from datasets import load_dataset
import re
import contractions
import emoji
import string
import random
from torch.utils.data import Dataset, DataLoader, TensorDataset, SequentialSampler
from sklearn.metrics import matthews_corrcoef





TRAIN_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/train_dataset.csv"
VAL_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/val_dataset.csv"
TEST_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/test_dataset.csv"
SUBMISSION_PATH = "/kaggle/working/submission.csv"

MODEL_NAME = "bert-base-uncased"  
BATCH_SIZE = 96  
EPOCHS = 2
LEARNING_RATE = 4e-5  
MAX_LEN = 40
RANDOM_SEED = 42
NUM_CLASSES = 2


ATTENTION_DROPOUT = 0.15      # Standard BERT attention dropout
HIDDEN_DROPOUT = 0.15       # Standard BERT hidden dropout
CLASSIFIER_DROPOUT = 0.15     # BERT classifier dropout
WEIGHT_DECAY = 0.065          # Standard weight decay for BERT

torch.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)



def reduce_repeated_letters(text):
    return re.sub(r"(.)\1{2,}", r"\1\1", text)


def clean_text(text):

    text = str(text).strip()
    
    # Only expand contractions and reduce repeated letters
    text = contractions.fix(text)
    text = reduce_repeated_letters(text)
    
    # Clean up extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text




train_df = pd.read_csv(TRAIN_PATH)
val_df = pd.read_csv(VAL_PATH)
test_df = pd.read_csv(TEST_PATH)

train_df["Text"] = train_df["Text"].apply(clean_text)
val_df["Text"] = val_df["Text"].apply(clean_text)
test_df["Text"] = test_df["Text"].apply(clean_text)


import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def analyze_text_lengths(train_df, val_df, test_df, tokenizer):

    
    # Function to get token length for a text
    def get_token_length(text):
        tokens = tokenizer.encode(text, add_special_tokens=True)
        return len(tokens)
    
    # Calculate token lengths for all datasets
    print("Calculating token lengths...")
    train_lengths = [get_token_length(text) for text in train_df["Text"]]
    val_lengths = [get_token_length(text) for text in val_df["Text"]]
    test_lengths = [get_token_length(text) for text in test_df["Text"]]
    
    # Combine all lengths
    all_lengths = train_lengths + val_lengths + test_lengths
    
    # Calculate statistics
    stats = {
        'min': np.min(all_lengths),
        'max': np.max(all_lengths),
        'mean': np.mean(all_lengths),
        'median': np.median(all_lengths),
        'std': np.std(all_lengths),
        'q25': np.percentile(all_lengths, 25),
        'q75': np.percentile(all_lengths, 75),
        'q90': np.percentile(all_lengths, 90),
        'q95': np.percentile(all_lengths, 95),
        'q99': np.percentile(all_lengths, 99)
    }
    
    print("\n TOKEN LENGTH STATISTICS:")
    print(f"Min length: {stats['min']}")
    print(f"Max length: {stats['max']}")
    print(f"Mean length: {stats['mean']:.2f}")
    print(f"Median length: {stats['median']:.2f}")
    print(f"Std deviation: {stats['std']:.2f}")
    print(f"25th percentile: {stats['q25']:.2f}")
    print(f"75th percentile: {stats['q75']:.2f}")
    print(f"90th percentile: {stats['q90']:.2f}")
    print(f"95th percentile: {stats['q95']:.2f}")
    print(f"99th percentile: {stats['q99']:.2f}")
    
    # Calculate coverage for different MAX_LEN values
    max_len_options = [32, 64, 80, 96, 128, 160, 192, 256, 320, 384, 512]
    coverage_data = []
    
    print("\n COVERAGE ANALYSIS FOR DIFFERENT MAX_LEN VALUES:")
    print("MAX_LEN | Coverage | Truncated | Avg Padding")
    print("-" * 45)
    
    for max_len in max_len_options:
        covered = sum(1 for length in all_lengths if length <= max_len)
        coverage_pct = (covered / len(all_lengths)) * 100
        truncated = len(all_lengths) - covered
        avg_padding = max_len - np.mean([min(length, max_len) for length in all_lengths])
        
        coverage_data.append({
            'max_len': max_len,
            'coverage': coverage_pct,
            'truncated': truncated,
            'avg_padding': avg_padding
        })
        
        print(f"{max_len:7d} | {coverage_pct:7.2f}% | {truncated:9d} | {avg_padding:11.2f}")
    
    # Visualization
    plt.figure(figsize=(20, 12))
    
    # 1. Distribution histogram
    plt.subplot(2, 3, 1)
    plt.hist(all_lengths, bins=50, alpha=0.7, edgecolor='black')
    plt.axvline(stats['mean'], color='red', linestyle='--', label=f'Mean: {stats["mean"]:.1f}')
    plt.axvline(stats['median'], color='green', linestyle='--', label=f'Median: {stats["median"]:.1f}')
    plt.axvline(stats['q95'], color='orange', linestyle='--', label=f'95th percentile: {stats["q95"]:.1f}')
    plt.xlabel('Token Length')
    plt.ylabel('Frequency')
    plt.title('Distribution of Token Lengths')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Box plot by dataset
    plt.subplot(2, 3, 2)
    data_for_box = [train_lengths, val_lengths, test_lengths]
    labels = ['Train', 'Validation', 'Test']
    plt.boxplot(data_for_box, labels=labels)
    plt.ylabel('Token Length')
    plt.title('Token Length Distribution by Dataset')
    plt.grid(True, alpha=0.3)
    
    # 3. Cumulative distribution
    plt.subplot(2, 3, 3)
    sorted_lengths = np.sort(all_lengths)
    percentiles = np.arange(1, len(sorted_lengths) + 1) / len(sorted_lengths) * 100
    plt.plot(sorted_lengths, percentiles)
    plt.xlabel('Token Length')
    plt.ylabel('Cumulative Percentage')
    plt.title('Cumulative Distribution of Token Lengths')
    plt.grid(True, alpha=0.3)
    
    # Add common MAX_LEN markers
    for max_len in [64, 80, 128, 256]:
        coverage = (np.searchsorted(sorted_lengths, max_len) / len(sorted_lengths)) * 100
        plt.axvline(max_len, color='red', linestyle='--', alpha=0.5)
        plt.text(max_len, coverage + 2, f'{max_len}', rotation=90, ha='center')
    
    # 4. Coverage vs MAX_LEN
    plt.subplot(2, 3, 4)
    coverages = [item['coverage'] for item in coverage_data]
    max_lens = [item['max_len'] for item in coverage_data]
    plt.plot(max_lens, coverages, marker='o', linewidth=2, markersize=8)
    plt.xlabel('MAX_LEN')
    plt.ylabel('Coverage (%)')
    plt.title('Coverage vs MAX_LEN')
    plt.grid(True, alpha=0.3)
    
    # Highlight good options
    for i, (max_len, coverage) in enumerate(zip(max_lens, coverages)):
        if coverage >= 95:
            plt.annotate(f'{max_len}\n({coverage:.1f}%)', 
                        (max_len, coverage), 
                        textcoords="offset points", 
                        xytext=(0,10), 
                        ha='center', 
                        fontsize=9,
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    # 5. Efficiency analysis (Coverage vs Padding)
    plt.subplot(2, 3, 5)
    avg_paddings = [item['avg_padding'] for item in coverage_data]
    plt.scatter(avg_paddings, coverages, s=100, alpha=0.7)
    
    for i, max_len in enumerate(max_lens):
        plt.annotate(str(max_len), (avg_paddings[i], coverages[i]), 
                    textcoords="offset points", xytext=(5,5), ha='left')
    
    plt.xlabel('Average Padding')
    plt.ylabel('Coverage (%)')
    plt.title('Efficiency: Coverage vs Average Padding')
    plt.grid(True, alpha=0.3)
    
    # 6. Length distribution by class (if labels available)
    plt.subplot(2, 3, 6)
    if 'Label' in train_df.columns:
        neg_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 0]
        pos_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 1]
        
        plt.hist(neg_lengths, bins=30, alpha=0.7, label='Negative', color='red')
        plt.hist(pos_lengths, bins=30, alpha=0.7, label='Positive', color='blue')
        plt.xlabel('Token Length')
        plt.ylabel('Frequency')
        plt.title('Token Length Distribution by Class')
        plt.legend()
        plt.grid(True, alpha=0.3)
    else:
        plt.text(0.5, 0.5, 'No labels available\nfor class analysis', 
                ha='center', va='center', transform=plt.gca().transAxes,
                fontsize=12, bbox=dict(boxstyle="round", facecolor='lightgray'))
        plt.title('Class Analysis Not Available')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/max_len_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Recommendations
    print("\nRECOMMENDATIONS:")
    print("=" * 40)
    
    # Find the most efficient MAX_LEN values
    efficient_options = []
    for item in coverage_data:
        if item['coverage'] >= 90:  # At least 90% coverage
            efficiency = item['coverage'] / (item['max_len'] / stats['mean'])  # Coverage per relative length
            efficient_options.append((item['max_len'], item['coverage'], efficiency))
    
    efficient_options.sort(key=lambda x: x[2], reverse=True)  # Sort by efficiency
    
    print("Most efficient MAX_LEN options (>=90% coverage):")
    for max_len, coverage, efficiency in efficient_options[:3]:
        print(f"  MAX_LEN = {max_len}: {coverage:.1f}% coverage (efficiency: {efficiency:.2f})")
    
    # Specific recommendations based on percentiles
    print(f"\nBased on percentiles:")
    print(f"  Conservative (99% coverage): MAX_LEN = {int(stats['q99']) + 5}")
    print(f"  Balanced (95% coverage): MAX_LEN = {int(stats['q95']) + 5}")
    print(f"  Aggressive (90% coverage): MAX_LEN = {int(stats['q90']) + 5}")
    print(f"  Very aggressive (75% coverage): MAX_LEN = {int(stats['q75']) + 5}")
    
    # Return the statistics for further use
    return stats, coverage_data

# Add this after loading and cleaning your data, but before creating datasets
print(" Starting EDA for MAX_LEN optimization...")
stats, coverage_data = analyze_text_lengths(train_df, val_df, test_df, tokenizer)

# Suggest optimal MAX_LEN based on analysis
optimal_max_len = int(stats['q95']) + 5  # 95th percentile + buffer
print(f"\nSUGGESTED MAX_LEN: {optimal_max_len}")


class TweetDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long)
        }


train_dataset = TweetDataset(train_df["Text"].tolist(), train_df["Label"].tolist(), tokenizer, MAX_LEN)
val_dataset = TweetDataset(val_df["Text"].tolist(), val_df["Label"].tolist(), tokenizer, MAX_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)


from transformers import AutoConfig

config = AutoConfig.from_pretrained(MODEL_NAME)

config.attention_probs_dropout_prob = ATTENTION_DROPOUT  # Attention dropout
config.hidden_dropout_prob = HIDDEN_DROPOUT              # Hidden layer dropout

if hasattr(config, 'classifier_dropout_prob'):
    config.classifier_dropout_prob = CLASSIFIER_DROPOUT
elif hasattr(config, 'classifier_dropout'):
    config.classifier_dropout = CLASSIFIER_DROPOUT
    

config.num_labels = NUM_CLASSES
config.id2label = {0: "negative", 1: "positive"}
config.label2id = {"negative": 0, "positive": 1}

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    config=config
).to(device)




def create_optimizer_with_weight_decay(model, learning_rate, weight_decay):

    # Separate parameters that should and shouldn't have weight decay
    decay_parameters = []
    no_decay_parameters = []
    
    for name, param in model.named_parameters():
        if not param.requires_grad:
            continue
            
        # Don't apply weight decay to bias terms and layer norm parameters
        if any(nd in name for nd in ["bias", "LayerNorm.weight", "layer_norm.weight"]):
            no_decay_parameters.append(param)
        else:
            decay_parameters.append(param)
    
    # Create parameter groups
    optimizer_grouped_parameters = [
        {
            "params": decay_parameters,
            "weight_decay": weight_decay,
            "lr": learning_rate
        },
        {
            "params": no_decay_parameters,
            "weight_decay": 0.0,
            "lr": learning_rate
        }
    ]
    

    return optim.AdamW(optimizer_grouped_parameters)

# Create enhanced optimizer
optimizer = create_optimizer_with_weight_decay(model, LEARNING_RATE, WEIGHT_DECAY)

# Learning rate scheduler - BERT typically benefits from warmup
total_steps = len(train_loader) * EPOCHS
warmup_steps = int(0.1 * total_steps)  # 10% warmup
scheduler = get_linear_schedule_with_warmup(optimizer, warmup_steps, total_steps)




def train_epoch(model, loader):
    model.train()
    total_loss = 0
    batch_losses = []
    all_preds = []
    all_labels = []
    all_probs = []
    
    for batch in tqdm(loader, desc="Training"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()
        batch_losses.append(loss.item())

        # Get predictions and track metrics during training
        logits = outputs.logits
        probs = F.softmax(logits, dim=1)
        preds = torch.argmax(logits, dim=1)
        
        # Collect for metrics calculation
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())
        all_probs.extend(probs[:, 1].cpu().tolist())

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()
        
    # Calculate metrics
    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass
    
    mcc = matthews_corrcoef(all_labels, all_preds)

    return avg_loss, batch_losses, acc, p, r, f, cm, auc_score, all_labels, all_probs, mcc


def evaluate(model, loader):
    model.eval()  
    all_preds = []
    all_labels = []
    all_probs = []
    total_loss = 0
    
    with torch.no_grad():
        for batch in tqdm(loader, desc="Evaluating"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()
            
            # Get predictions and probabilities
            logits = outputs.logits
            probs = F.softmax(logits, dim=1)
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())
            all_probs.extend(probs[:, 1].cpu().tolist())

    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass

    mcc = matthews_corrcoef(all_labels, all_preds)
    
    return acc, p, r, f, avg_loss, cm, auc_score, all_labels, all_probs, mcc


def plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, 
                 model_name, batch_losses=None):
    """Plot training and validation metrics"""
    plt.figure(figsize=(15, 10))
    
    # Plot losses
    plt.subplot(2, 2, 1)
    plt.plot(train_losses, label="Train Loss", marker='o')
    plt.plot(val_losses, label="Validation Loss", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid(True)
    
    # Plot accuracy
    plt.subplot(2, 2, 2)
    plt.plot(train_accs, label="Train Accuracy", marker='o')
    plt.plot(val_accs, label="Validation Accuracy", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Training and Validation Accuracy")
    plt.legend()
    plt.grid(True)
    
    # Plot F1 score
    plt.subplot(2, 2, 3)
    plt.plot(train_f1s, label="Train F1", marker='o')
    plt.plot(val_f1s, label="Validation F1", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("F1 Score")
    plt.title("Training and Validation F1 Score")
    plt.legend()
    plt.grid(True)
    
    # Plot batch-level losses if available
    if batch_losses:
        plt.subplot(2, 2, 4)
        all_batches = []
        for epoch, losses in enumerate(batch_losses):
            batches = list(range(epoch * len(losses), (epoch + 1) * len(losses)))
            all_batches.extend(batches)
            plt.plot(batches, losses, alpha=0.5, label=f"Epoch {epoch+1}")
        
        plt.xlabel("Batch")
        plt.ylabel("Loss")
        plt.title("Batch-level Loss")
        plt.legend()
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig(f"/kaggle/working/{model_name}_learning_curves.png")
    plt.show()

def plot_confusion_matrix(cm, model_name, set_name="Validation"):
    """Plot confusion matrix"""
    plt.figure(figsize=(8, 6))
    display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
    display.plot(cmap=plt.cm.Blues, values_format=".0f")
    plt.title(f"{set_name} Confusion Matrix")
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_confusion_matrix.png")
    plt.show()

def plot_roc_curve(y_true, y_probs, model_name, set_name="Validation"):
    """Plot ROC curve"""
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{set_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_roc_curve.png")
    plt.show()
    
    return roc_auc


best_f1 = 0
model_path = f"/kaggle/working/best_{MODEL_NAME.replace('-', '_')}.pth"

# Metrics tracking
train_losses = []
val_losses = []
train_accs = []
val_accs = []
train_f1s = []
val_f1s = []
train_aucs = []
val_aucs = []
batch_level_losses = []

for epoch in range(EPOCHS):
    print(f"\nEpoch {epoch + 1}/{EPOCHS}")
    
    # Train and collect metrics directly during training
    train_loss, batch_losses, train_acc, train_p, train_r, train_f1, train_cm, train_auc, train_labels, train_probs, train_mcc = train_epoch(model, train_loader)
    batch_level_losses.append(batch_losses)
    
    # Evaluate only on validation set
    val_acc, val_p, val_r, val_f1, val_loss, val_cm, val_auc, val_labels, val_probs, val_mcc = evaluate(model, val_loader)

    # Store metrics
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    train_f1s.append(train_f1)
    val_f1s.append(val_f1)
    
    # Add AUC scores if available
    if train_auc is not None:
        train_aucs.append(train_auc)
    if val_auc is not None:
        val_aucs.append(val_auc)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Train  - Acc: {train_acc:.4f}, Precision: {train_p:.4f}, Recall: {train_r:.4f}, F1: {train_f1:.4f}")
    if train_auc is not None:
        print(f"         AUC: {train_auc:.4f}")
    
    print(f"Val    - Acc: {val_acc:.4f}, Precision: {val_p:.4f}, Recall: {val_r:.4f}, F1: {val_f1:.4f}, Loss: {val_loss:.4f}")
    if val_auc is not None:
        print(f"         AUC: {val_auc:.4f}")

    # Save best model
    if val_f1 > best_f1:
        best_f1 = val_f1
        torch.save(model.state_dict(), model_path)
        print(" Best model saved!")
        
        # Plot confusion matrix and ROC curve for best model
        plot_confusion_matrix(val_cm, MODEL_NAME)
        if val_auc is not None:
            plot_roc_curve(val_labels, val_probs, MODEL_NAME)


plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, MODEL_NAME, batch_level_losses)


# Load the best model
model.load_state_dict(torch.load(model_path))
model.eval()

# Evaluate the best model on validation set for final metrics
final_val_acc, final_val_p, final_val_r, final_val_f1, final_val_loss, final_val_cm, final_val_auc, final_val_labels, final_val_probs, final_val_mcc = evaluate(model, val_loader)
print("\n===== FINAL MODEL EVALUATION =====")
print(f"Val    - Acc: {final_val_acc:.4f}, Precision: {final_val_p:.4f}, Recall: {final_val_r:.4f}, F1: {final_val_f1:.4f}, MCC: {final_val_mcc:.4f}, Loss: {final_val_loss:.4f}")
if final_val_auc is not None:
    print(f"         AUC: {final_val_auc:.4f}")


# Plot final confusion matrix and ROC curve
plot_confusion_matrix(final_val_cm, MODEL_NAME, "Validation")
if final_val_auc is not None:
    plot_roc_curve(final_val_labels, final_val_probs, MODEL_NAME, "Final Validation")


class TestDataset(Dataset):
    def __init__(self, texts, tokenizer, max_len):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0)
        }

test_dataset = TestDataset(test_df["Text"].tolist(), tokenizer, MAX_LEN)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)

test_preds = []
with torch.no_grad():
    for batch in tqdm(test_loader, desc="Predicting"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1)
        
        test_preds.extend(preds.cpu().tolist())


id_col = "ID" if "ID" in test_df.columns else "Id"
submission = pd.DataFrame({id_col: test_df[id_col], "Label": test_preds})
submission.to_csv(SUBMISSION_PATH, index=False)
print(f"Submission saved to: {SUBMISSION_PATH}")
```

```python


!pip install scikit-learn

!pip install transformers

!pip install emoji contractions

!pip install datasets



import pandas as pd

import numpy as np

import torch

import torch.nn.functional as F

import torch.optim as optim

from torch.utils.data import Dataset, DataLoader


from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_linear_schedule_with_warmup


from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay

from sklearn.metrics import roc_curve, auc, roc_auc_score

import matplotlib.pyplot as plt

import seaborn as sns

from tqdm import tqdm

from datasets import load_dataset

import re

import contractions

import emoji

import string

import random

from torch.utils.data import Dataset, DataLoader, TensorDataset, SequentialSampler

from sklearn.metrics import matthews_corrcoef



TRAIN_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/train_dataset.csv"

VAL_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/val_dataset.csv"

TEST_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/test_dataset.csv"

SUBMISSION_PATH = "/kaggle/working/submission.csv"


MODEL_NAME = "bert-base-uncased"  

BATCH_SIZE = 96

EPOCHS = 3  

LEARNING_RATE = 4e-5  

MAX_LEN = 40  

RANDOM_SEED = 42

NUM_CLASSES = 2


torch.manual_seed(RANDOM_SEED)

np.random.seed(RANDOM_SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")



tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)



def reduce_repeated_letters(text):

    return re.sub(r"(.)\1{2,}", r"\1\1", text)
    

def clean_text(text):

    text = str(text).strip()
    
    
    # Only expand contractions and reduce repeated letters
    
    text = contractions.fix(text)
    
    text = reduce_repeated_letters(text)
    
    
    # Clean up extra whitespace
    
    text = re.sub(r'\s+', ' ', text).strip()
    
    
    return text
    

train_df = pd.read_csv(TRAIN_PATH)

val_df = pd.read_csv(VAL_PATH)

test_df = pd.read_csv(TEST_PATH)


train_df["Text"] = train_df["Text"].apply(clean_text)

val_df["Text"] = val_df["Text"].apply(clean_text)

test_df["Text"] = test_df["Text"].apply(clean_text)



import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np


def analyze_text_lengths(train_df, val_df, test_df, tokenizer):
    
    """
    Analyze text lengths to determine optimal MAX_LEN for BERT
    """
    print("ANALYZING TEXT LENGTHS FOR BERT MAX_LEN OPTIMIZATION")
    print("=" * 60)
    
    # Function to get token length for a text
    def get_token_length(text):
        tokens = tokenizer.encode(text, add_special_tokens=True)
        return len(tokens)
    
    # Calculate token lengths for all datasets
    print("Calculating token lengths...")
    train_lengths = [get_token_length(text) for text in train_df["Text"]]
    val_lengths = [get_token_length(text) for text in val_df["Text"]]
    test_lengths = [get_token_length(text) for text in test_df["Text"]]
    
    # Combine all lengths
    all_lengths = train_lengths + val_lengths + test_lengths
    
    # Calculate statistics
    stats = {
        'min': np.min(all_lengths),
        'max': np.max(all_lengths),
        'mean': np.mean(all_lengths),
        'median': np.median(all_lengths),
        'std': np.std(all_lengths),
        'q25': np.percentile(all_lengths, 25),
        'q75': np.percentile(all_lengths, 75),
        'q90': np.percentile(all_lengths, 90),
        'q95': np.percentile(all_lengths, 95),
        'q99': np.percentile(all_lengths, 99)
    }
    
    print("\nTOKEN LENGTH STATISTICS:")
    print(f"Min length: {stats['min']}")
    print(f"Max length: {stats['max']}")
    print(f"Mean length: {stats['mean']:.2f}")
    print(f"Median length: {stats['median']:.2f}")
    print(f"Std deviation: {stats['std']:.2f}")
    print(f"25th percentile: {stats['q25']:.2f}")
    print(f"75th percentile: {stats['q75']:.2f}")
    print(f"90th percentile: {stats['q90']:.2f}")
    print(f"95th percentile: {stats['q95']:.2f}")
    print(f"99th percentile: {stats['q99']:.2f}")
    
    # Calculate coverage for different MAX_LEN values (BERT optimized)
    max_len_options = [64, 96, 128, 160, 192, 256, 320, 384, 512]  # BERT can handle up to 512
    coverage_data = []
    
    print("\n📊 COVERAGE ANALYSIS FOR DIFFERENT MAX_LEN VALUES (BERT):")
    print("MAX_LEN | Coverage | Truncated | Avg Padding")
    print("-" * 45)
    
    for max_len in max_len_options:
        covered = sum(1 for length in all_lengths if length <= max_len)
        coverage_pct = (covered / len(all_lengths)) * 100
        truncated = len(all_lengths) - covered
        avg_padding = max_len - np.mean([min(length, max_len) for length in all_lengths])
        
        coverage_data.append({
            'max_len': max_len,
            'coverage': coverage_pct,
            'truncated': truncated,
            'avg_padding': avg_padding
        })
        
        print(f"{max_len:7d} | {coverage_pct:7.2f}% | {truncated:9d} | {avg_padding:11.2f}")
    
    # Visualization
    plt.figure(figsize=(20, 12))
    
    # 1. Distribution histogram
    plt.subplot(2, 3, 1)
    plt.hist(all_lengths, bins=50, alpha=0.7, edgecolor='black')
    plt.axvline(stats['mean'], color='red', linestyle='--', label=f'Mean: {stats["mean"]:.1f}')
    plt.axvline(stats['median'], color='green', linestyle='--', label=f'Median: {stats["median"]:.1f}')
    plt.axvline(stats['q95'], color='orange', linestyle='--', label=f'95th percentile: {stats["q95"]:.1f}')
    plt.xlabel('Token Length')
    plt.ylabel('Frequency')
    plt.title('Distribution of Token Lengths (BERT)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Box plot by dataset
    plt.subplot(2, 3, 2)
    data_for_box = [train_lengths, val_lengths, test_lengths]
    labels = ['Train', 'Validation', 'Test']
    plt.boxplot(data_for_box, labels=labels)
    plt.ylabel('Token Length')
    plt.title('Token Length Distribution by Dataset (BERT)')
    plt.grid(True, alpha=0.3)
    
    # 3. Cumulative distribution
    plt.subplot(2, 3, 3)
    sorted_lengths = np.sort(all_lengths)
    percentiles = np.arange(1, len(sorted_lengths) + 1) / len(sorted_lengths) * 100
    plt.plot(sorted_lengths, percentiles)
    plt.xlabel('Token Length')
    plt.ylabel('Cumulative Percentage')
    plt.title('Cumulative Distribution of Token Lengths (BERT)')
    plt.grid(True, alpha=0.3)
    
    # Add common MAX_LEN markers for BERT
    for max_len in [128, 192, 256, 512]:
        coverage = (np.searchsorted(sorted_lengths, max_len) / len(sorted_lengths)) * 100
        plt.axvline(max_len, color='red', linestyle='--', alpha=0.5)
        plt.text(max_len, coverage + 2, f'{max_len}', rotation=90, ha='center')
    
    # 4. Coverage vs MAX_LEN
    plt.subplot(2, 3, 4)
    coverages = [item['coverage'] for item in coverage_data]
    max_lens = [item['max_len'] for item in coverage_data]
    plt.plot(max_lens, coverages, marker='o', linewidth=2, markersize=8)
    plt.xlabel('MAX_LEN')
    plt.ylabel('Coverage (%)')
    plt.title('Coverage vs MAX_LEN (BERT)')
    plt.grid(True, alpha=0.3)
    
    # Highlight good options
    for i, (max_len, coverage) in enumerate(zip(max_lens, coverages)):
        if coverage >= 95:
            plt.annotate(f'{max_len}\n({coverage:.1f}%)', 
                        (max_len, coverage), 
                        textcoords="offset points", 
                        xytext=(0,10), 
                        ha='center', 
                        fontsize=9,
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    # 5. Efficiency analysis (Coverage vs Padding)
    plt.subplot(2, 3, 5)
    avg_paddings = [item['avg_padding'] for item in coverage_data]
    plt.scatter(avg_paddings, coverages, s=100, alpha=0.7)
    
    for i, max_len in enumerate(max_lens):
        plt.annotate(str(max_len), (avg_paddings[i], coverages[i]), 
                    textcoords="offset points", xytext=(5,5), ha='left')
    
    plt.xlabel('Average Padding')
    plt.ylabel('Coverage (%)')
    plt.title('Efficiency: Coverage vs Average Padding (BERT)')
    plt.grid(True, alpha=0.3)
    
    # 6. Length distribution by class (if labels available)
    plt.subplot(2, 3, 6)
    if 'Label' in train_df.columns:
        neg_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 0]
        pos_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 1]
        
        plt.hist(neg_lengths, bins=30, alpha=0.7, label='Negative', color='red')
        plt.hist(pos_lengths, bins=30, alpha=0.7, label='Positive', color='blue')
        plt.xlabel('Token Length')
        plt.ylabel('Frequency')
        plt.title('Token Length Distribution by Class (BERT)')
        plt.legend()
        plt.grid(True, alpha=0.3)
    else:
        plt.text(0.5, 0.5, 'No labels available\nfor class analysis', 
                ha='center', va='center', transform=plt.gca().transAxes,
                fontsize=12, bbox=dict(boxstyle="round", facecolor='lightgray'))
        plt.title('Class Analysis Not Available')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/bert_max_len_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # BERT-specific recommendations
    print("\n BERT-SPECIFIC RECOMMENDATIONS:")
    print("=" * 40)
    
    # Find the most efficient MAX_LEN values for BERT
    efficient_options = []
    for item in coverage_data:
        if item['coverage'] >= 90:  # At least 90% coverage
            efficiency = item['coverage'] / (item['max_len'] / stats['mean'])  # Coverage per relative length
            efficient_options.append((item['max_len'], item['coverage'], efficiency))
    
    efficient_options.sort(key=lambda x: x[2], reverse=True)  # Sort by efficiency
    
    print("Most efficient MAX_LEN options for BERT (>=90% coverage):")
    for max_len, coverage, efficiency in efficient_options[:3]:
        print(f"  MAX_LEN = {max_len}: {coverage:.1f}% coverage (efficiency: {efficiency:.2f})")
    
    # BERT-specific recommendations based on percentiles
    print(f"\nBased on percentiles (BERT optimized):")
    print(f"  Conservative (99% coverage): MAX_LEN = {min(int(stats['q99']) + 5, 512)}")
    print(f"  Balanced (95% coverage): MAX_LEN = {min(int(stats['q95']) + 5, 256)}")
    print(f"  Aggressive (90% coverage): MAX_LEN = {min(int(stats['q90']) + 5, 192)}")
    print(f"  Very aggressive (75% coverage): MAX_LEN = {min(int(stats['q75']) + 5, 128)}")

    
    # Return the statistics for further use
    return stats, coverage_data


print("Starting EDA for BERT MAX_LEN optimization...")


stats, coverage_data = analyze_text_lengths(train_df, val_df, test_df, tokenizer)



optimal_max_len = min(int(stats['q95']) + 5, 256)  





class TweetDataset(Dataset):

    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long)
        }


train_dataset = TweetDataset(train_df["Text"].tolist(), train_df["Label"].tolist(), tokenizer, MAX_LEN)


val_dataset = TweetDataset(val_df["Text"].tolist(), val_df["Label"].tolist(), tokenizer, MAX_LEN)



train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)


val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)






model = AutoModelForSequenceClassification.from_pretrained(
   
    MODEL_NAME,
    num_labels=NUM_CLASSES,
    id2label={0: "negative", 1: "positive"},
    label2id={"negative": 0, "positive": 1}
).to(device)


optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=0.01)  # Added weight decay


total_steps = len(train_loader) * EPOCHS


warmup_steps = int(0.1 * total_steps)  # 10% warmup


scheduler = get_linear_schedule_with_warmup(optimizer, warmup_steps, total_steps)






def train_epoch(model, loader):

    model.train()
    total_loss = 0
    batch_losses = []  # Track loss per batch
    all_preds = []
    all_labels = []
    all_probs = []  # Store probabilities for training metrics
    
    for batch in tqdm(loader, desc="Training"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()
        batch_losses.append(loss.item())

        # Get predictions and track metrics during training
        logits = outputs.logits
        probs = F.softmax(logits, dim=1)
        preds = torch.argmax(logits, dim=1)
        
        # Collect for metrics calculation
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())
        all_probs.extend(probs[:, 1].cpu().tolist())  # Probability of positive class

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)  # Gradient clipping for BERT
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()
        
    # Calculate metrics
    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass  # In case of single class
    
    mcc = matthews_corrcoef(all_labels, all_preds)

    return avg_loss, batch_losses, acc, p, r, f, cm, auc_score, all_labels, all_probs, mcc


def evaluate(model, loader):

    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []  # Store probabilities for ROC curve
    total_loss = 0
    
    with torch.no_grad():
        for batch in tqdm(loader, desc="Evaluating"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()
            
            # Get predictions and probabilities
            logits = outputs.logits
            probs = F.softmax(logits, dim=1)
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())
            all_probs.extend(probs[:, 1].cpu().tolist())  # Probability of positive class

    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    
    # Create confusion matrix
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass  # In case of single class

    mcc = matthews_corrcoef(all_labels, all_preds)

    return acc, p, r, f, avg_loss, cm, auc_score, all_labels, all_probs, mcc

def plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, 
                 model_name, batch_losses=None):
                 
    """Plot training and validation metrics"""
    plt.figure(figsize=(15, 10))
    
    # Plot losses
    plt.subplot(2, 2, 1)
    plt.plot(train_losses, label="Train Loss", marker='o')
    plt.plot(val_losses, label="Validation Loss", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid(True)
    
    # Plot accuracy
    plt.subplot(2, 2, 2)
    plt.plot(train_accs, label="Train Accuracy", marker='o')
    plt.plot(val_accs, label="Validation Accuracy", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Training and Validation Accuracy")
    plt.legend()
    plt.grid(True)
    
    # Plot F1 score
    plt.subplot(2, 2, 3)
    plt.plot(train_f1s, label="Train F1", marker='o')
    plt.plot(val_f1s, label="Validation F1", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("F1 Score")
    plt.title("Training and Validation F1 Score")
    plt.legend()
    plt.grid(True)
    
    # Plot batch-level losses if available
    if batch_losses:
        plt.subplot(2, 2, 4)
        all_batches = []
        for epoch, losses in enumerate(batch_losses):
            batches = list(range(epoch * len(losses), (epoch + 1) * len(losses)))
            all_batches.extend(batches)
            plt.plot(batches, losses, alpha=0.5, label=f"Epoch {epoch+1}")
        
        plt.xlabel("Batch")
        plt.ylabel("Loss")
        plt.title("Batch-level Loss")
        plt.legend()
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig(f"/kaggle/working/{model_name}_learning_curves.png")
    plt.show()

def plot_confusion_matrix(cm, model_name, set_name="Validation"):

    """Plot confusion matrix"""
    plt.figure(figsize=(8, 6))
    display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
    display.plot(cmap=plt.cm.Blues, values_format=".0f")
    plt.title(f"{set_name} Confusion Matrix")
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_confusion_matrix.png")
    plt.show()

def plot_roc_curve(y_true, y_probs, model_name, set_name="Validation"):

    """Plot ROC curve"""
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{set_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_roc_curve.png")
    plt.show()
    
    return roc_auc

best_f1 = 0

model_path = f"/kaggle/working/best_{MODEL_NAME.replace('-', '_')}.pth"


train_losses = []
val_losses = []
train_accs = []
val_accs = []
train_f1s = []
val_f1s = []
train_aucs = []
val_aucs = []
batch_level_losses = []

for epoch in range(EPOCHS):

    print(f"\nEpoch {epoch + 1}/{EPOCHS}")
    
    # Train and collect metrics directly during training
    train_loss, batch_losses, train_acc, train_p, train_r, train_f1, train_cm, train_auc, train_labels, train_probs, train_mcc = train_epoch(model, train_loader)
    batch_level_losses.append(batch_losses)
    
    # Evaluate only on validation set
    val_acc, val_p, val_r, val_f1, val_loss, val_cm, val_auc, val_labels, val_probs, val_mcc = evaluate(model, val_loader)

    # Store metrics
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    train_f1s.append(train_f1)
    val_f1s.append(val_f1)
    
    # Add AUC scores if available
    if train_auc is not None:
        train_aucs.append(train_auc)
    if val_auc is not None:
        val_aucs.append(val_auc)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Train  - Acc: {train_acc:.4f}, Precision: {train_p:.4f}, Recall: {train_r:.4f}, F1: {train_f1:.4f}")
    if train_auc is not None:
        print(f"         AUC: {train_auc:.4f}")
    
    print(f"Val    - Acc: {val_acc:.4f}, Precision: {val_p:.4f}, Recall: {val_r:.4f}, F1: {val_f1:.4f}, Loss: {val_loss:.4f}")
    if val_auc is not None:
        print(f"         AUC: {val_auc:.4f}")

    # Save best model
    if val_f1 > best_f1:
        best_f1 = val_f1
        torch.save(model.state_dict(), model_path)
        print(" Best model saved!")
        
        # Plot confusion matrix and ROC curve for best model
        plot_confusion_matrix(val_cm, MODEL_NAME)
        if val_auc is not None:
            plot_roc_curve(val_labels, val_probs, MODEL_NAME)


plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, MODEL_NAME, batch_level_losses)



model.load_state_dict(torch.load(model_path))

model.eval()


final_val_acc, final_val_p, final_val_r, final_val_f1, final_val_loss, final_val_cm, final_val_auc, final_val_labels, final_val_probs, final_val_mcc = evaluate(model, val_loader)

print("\n===== FINAL MODEL EVALUATION =====")

print(f"Val    - Acc: {final_val_acc:.4f}, Precision: {final_val_p:.4f}, Recall: {final_val_r:.4f}, F1: {final_val_f1:.4f}, MCC: {final_val_mcc:.4f}, Loss: {final_val_loss:.4f}")

if final_val_auc is not None:

    print(f"         AUC: {final_val_auc:.4f}")



plot_confusion_matrix(final_val_cm, MODEL_NAME, "Validation")

if final_val_auc is not None:

    plot_roc_curve(final_val_labels, final_val_probs, MODEL_NAME, "Final Validation")


class TestDataset(Dataset):

    def __init__(self, texts, tokenizer, max_len):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0)
        }

test_dataset = TestDataset(test_df["Text"].tolist(), tokenizer, MAX_LEN)

test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)


test_preds = []

with torch.no_grad():

    for batch in tqdm(test_loader, desc="Predicting"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1)
        
        test_preds.extend(preds.cpu().tolist())


id_col = "ID" if "ID" in test_df.columns else "Id"

submission = pd.DataFrame({id_col: test_df[id_col], "Label": test_preds})

submission.to_csv(SUBMISSION_PATH, index=False)

print(f" Submission saved to: {SUBMISSION_PATH}")
```

```python
!pip install scikit-learn

!pip install transformers

!pip install emoji contractions

!pip install datasets

!pip install optuna



import pandas as pd

import numpy as np

import torch

import torch.nn.functional as F

import torch.optim as optim

from torch.utils.data import Dataset, DataLoader


from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_linear_schedule_with_warmup


from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay

from sklearn.metrics import roc_curve, auc, roc_auc_score

import matplotlib.pyplot as plt

import seaborn as sns

from tqdm import tqdm

from datasets import load_dataset

import re

import contractions

import emoji

import string

import random

from torch.utils.data import Dataset, DataLoader, TensorDataset, SequentialSampler
from sklearn.metrics import matthews_corrcoef

import optuna

from transformers import AutoConfig



TRAIN_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/train_dataset.csv"

VAL_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/val_dataset.csv"

TEST_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/test_dataset.csv"

SUBMISSION_PATH = "/kaggle/working/submission.csv"


MODEL_NAME = "bert-base-uncased"  

BATCH_SIZE = 96  

EPOCHS = 3

LEARNING_RATE = 4e-5 

MAX_LEN = 40

RANDOM_SEED = 42

NUM_CLASSES = 2


torch.manual_seed(RANDOM_SEED)

np.random.seed(RANDOM_SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")



tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)



def reduce_repeated_letters(text):

    return re.sub(r"(.)\1{2,}", r"\1\1", text)


def clean_text(text):

    text = str(text).strip()
    
    # Only expand contractions and reduce repeated letters
    text = contractions.fix(text)
    text = reduce_repeated_letters(text)
    
    # Clean up extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

train_df = pd.read_csv(TRAIN_PATH)

val_df = pd.read_csv(VAL_PATH)

test_df = pd.read_csv(TEST_PATH)


train_df["Text"] = train_df["Text"].apply(clean_text)

val_df["Text"] = val_df["Text"].apply(clean_text)

test_df["Text"] = test_df["Text"].apply(clean_text)



import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np


def analyze_text_lengths(train_df, val_df, test_df, tokenizer):

    print(" ANALYZING TEXT LENGTHS FOR MAX_LEN OPTIMIZATION")
    print("=" * 60)
    
    # Function to get token length for a text
    def get_token_length(text):
        tokens = tokenizer.encode(text, add_special_tokens=True)
        return len(tokens)
    
    # Calculate token lengths for all datasets
    print(" Calculating token lengths...")
    train_lengths = [get_token_length(text) for text in train_df["Text"]]
    val_lengths = [get_token_length(text) for text in val_df["Text"]]
    test_lengths = [get_token_length(text) for text in test_df["Text"]]
    
    # Combine all lengths
    all_lengths = train_lengths + val_lengths + test_lengths
    
    # Calculate statistics
    stats = {
        'min': np.min(all_lengths),
        'max': np.max(all_lengths),
        'mean': np.mean(all_lengths),
        'median': np.median(all_lengths),
        'std': np.std(all_lengths),
        'q25': np.percentile(all_lengths, 25),
        'q75': np.percentile(all_lengths, 75),
        'q90': np.percentile(all_lengths, 90),
        'q95': np.percentile(all_lengths, 95),
        'q99': np.percentile(all_lengths, 99)
    }
    
    print("\n TOKEN LENGTH STATISTICS:")
    print(f"Min length: {stats['min']}")
    print(f"Max length: {stats['max']}")
    print(f"Mean length: {stats['mean']:.2f}")
    print(f"Median length: {stats['median']:.2f}")
    print(f"Std deviation: {stats['std']:.2f}")
    print(f"25th percentile: {stats['q25']:.2f}")
    print(f"75th percentile: {stats['q75']:.2f}")
    print(f"90th percentile: {stats['q90']:.2f}")
    print(f"95th percentile: {stats['q95']:.2f}")
    print(f"99th percentile: {stats['q99']:.2f}")
    
    # Calculate coverage for different MAX_LEN values
    max_len_options = [32, 64, 80, 96, 128, 160, 192, 256, 320, 384, 512]
    coverage_data = []
    
    print("\n COVERAGE ANALYSIS FOR DIFFERENT MAX_LEN VALUES:")
    print("MAX_LEN | Coverage | Truncated | Avg Padding")
    print("-" * 45)
    
    for max_len in max_len_options:
        covered = sum(1 for length in all_lengths if length <= max_len)
        coverage_pct = (covered / len(all_lengths)) * 100
        truncated = len(all_lengths) - covered
        avg_padding = max_len - np.mean([min(length, max_len) for length in all_lengths])
        
        coverage_data.append({
            'max_len': max_len,
            'coverage': coverage_pct,
            'truncated': truncated,
            'avg_padding': avg_padding
        })
        
        print(f"{max_len:7d} | {coverage_pct:7.2f}% | {truncated:9d} | {avg_padding:11.2f}")
    
    # Visualization
    plt.figure(figsize=(20, 12))
    
    # 1. Distribution histogram
    plt.subplot(2, 3, 1)
    plt.hist(all_lengths, bins=50, alpha=0.7, edgecolor='black')
    plt.axvline(stats['mean'], color='red', linestyle='--', label=f'Mean: {stats["mean"]:.1f}')
    plt.axvline(stats['median'], color='green', linestyle='--', label=f'Median: {stats["median"]:.1f}')
    plt.axvline(stats['q95'], color='orange', linestyle='--', label=f'95th percentile: {stats["q95"]:.1f}')
    plt.xlabel('Token Length')
    plt.ylabel('Frequency')
    plt.title('Distribution of Token Lengths')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Box plot by dataset
    plt.subplot(2, 3, 2)
    data_for_box = [train_lengths, val_lengths, test_lengths]
    labels = ['Train', 'Validation', 'Test']
    plt.boxplot(data_for_box, labels=labels)
    plt.ylabel('Token Length')
    plt.title('Token Length Distribution by Dataset')
    plt.grid(True, alpha=0.3)
    
    # 3. Cumulative distribution
    plt.subplot(2, 3, 3)
    sorted_lengths = np.sort(all_lengths)
    percentiles = np.arange(1, len(sorted_lengths) + 1) / len(sorted_lengths) * 100
    plt.plot(sorted_lengths, percentiles)
    plt.xlabel('Token Length')
    plt.ylabel('Cumulative Percentage')
    plt.title('Cumulative Distribution of Token Lengths')
    plt.grid(True, alpha=0.3)
    
    # Add common MAX_LEN markers
    for max_len in [64, 80, 128, 256]:
        coverage = (np.searchsorted(sorted_lengths, max_len) / len(sorted_lengths)) * 100
        plt.axvline(max_len, color='red', linestyle='--', alpha=0.5)
        plt.text(max_len, coverage + 2, f'{max_len}', rotation=90, ha='center')
    
    # 4. Coverage vs MAX_LEN
    plt.subplot(2, 3, 4)
    coverages = [item['coverage'] for item in coverage_data]
    max_lens = [item['max_len'] for item in coverage_data]
    plt.plot(max_lens, coverages, marker='o', linewidth=2, markersize=8)
    plt.xlabel('MAX_LEN')
    plt.ylabel('Coverage (%)')
    plt.title('Coverage vs MAX_LEN')
    plt.grid(True, alpha=0.3)
    
    # Highlight good options
    for i, (max_len, coverage) in enumerate(zip(max_lens, coverages)):
        if coverage >= 95:
            plt.annotate(f'{max_len}\n({coverage:.1f}%)', 
                        (max_len, coverage), 
                        textcoords="offset points", 
                        xytext=(0,10), 
                        ha='center', 
                        fontsize=9,
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    # 5. Efficiency analysis (Coverage vs Padding)
    plt.subplot(2, 3, 5)
    avg_paddings = [item['avg_padding'] for item in coverage_data]
    plt.scatter(avg_paddings, coverages, s=100, alpha=0.7)
    
    for i, max_len in enumerate(max_lens):
        plt.annotate(str(max_len), (avg_paddings[i], coverages[i]), 
                    textcoords="offset points", xytext=(5,5), ha='left')
    
    plt.xlabel('Average Padding')
    plt.ylabel('Coverage (%)')
    plt.title('Efficiency: Coverage vs Average Padding')
    plt.grid(True, alpha=0.3)
    
    # 6. Length distribution by class (if labels available)
    plt.subplot(2, 3, 6)
    if 'Label' in train_df.columns:
        neg_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 0]
        pos_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 1]
        
        plt.hist(neg_lengths, bins=30, alpha=0.7, label='Negative', color='red')
        plt.hist(pos_lengths, bins=30, alpha=0.7, label='Positive', color='blue')
        plt.xlabel('Token Length')
        plt.ylabel('Frequency')
        plt.title('Token Length Distribution by Class')
        plt.legend()
        plt.grid(True, alpha=0.3)
    else:
        plt.text(0.5, 0.5, 'No labels available\nfor class analysis', 
                ha='center', va='center', transform=plt.gca().transAxes,
                fontsize=12, bbox=dict(boxstyle="round", facecolor='lightgray'))
        plt.title('Class Analysis Not Available')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/max_len_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Recommendations
    print("\n RECOMMENDATIONS:")
    print("=" * 40)
    
    # Find the most efficient MAX_LEN values
    efficient_options = []
    for item in coverage_data:
        if item['coverage'] >= 90:  # At least 90% coverage
            efficiency = item['coverage'] / (item['max_len'] / stats['mean'])  # Coverage per relative length
            efficient_options.append((item['max_len'], item['coverage'], efficiency))
    
    efficient_options.sort(key=lambda x: x[2], reverse=True)  # Sort by efficiency
    
    print("Most efficient MAX_LEN options (>=90% coverage):")
    for max_len, coverage, efficiency in efficient_options[:3]:
        print(f"  MAX_LEN = {max_len}: {coverage:.1f}% coverage (efficiency: {efficiency:.2f})")
    
    # Specific recommendations based on percentiles
    print(f"\nBased on percentiles:")
    print(f"  Conservative (99% coverage): MAX_LEN = {int(stats['q99']) + 5}")
    print(f"  Balanced (95% coverage): MAX_LEN = {int(stats['q95']) + 5}")
    print(f"  Aggressive (90% coverage): MAX_LEN = {int(stats['q90']) + 5}")
    print(f"  Very aggressive (75% coverage): MAX_LEN = {int(stats['q75']) + 5}")
    
    # Return the statistics for further use
    return stats, coverage_data


stats, coverage_data = analyze_text_lengths(train_df, val_df, test_df, tokenizer)


optimal_max_len = int(stats['q95']) + 5  # 95th percentile + buffer

print(f"\n SUGGESTED MAX_LEN: {optimal_max_len}")



class TweetDataset(Dataset):

    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long)
        }


train_dataset = TweetDataset(train_df["Text"].tolist(), train_df["Label"].tolist(), tokenizer, MAX_LEN)

val_dataset = TweetDataset(val_df["Text"].tolist(), val_df["Label"].tolist(), tokenizer, MAX_LEN)


train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)

val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)




def train_epoch(model, loader, optimizer, scheduler):

    model.train()
    total_loss = 0
    batch_losses = []  # Track loss per batch
    all_preds = []
    all_labels = []
    all_probs = []  # Store probabilities for training metrics
    
    for batch in tqdm(loader, desc="Training"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()
        batch_losses.append(loss.item())

        # Get predictions and track metrics during training
        logits = outputs.logits
        probs = F.softmax(logits, dim=1)
        preds = torch.argmax(logits, dim=1)
        
        # Collect for metrics calculation
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())
        all_probs.extend(probs[:, 1].cpu().tolist())  # Probability of positive class

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()
        
    # Calculate metrics
    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass  # In case of single class
    
    mcc = matthews_corrcoef(all_labels, all_preds)

    return avg_loss, batch_losses, acc, p, r, f, cm, auc_score, all_labels, all_probs, mcc



def evaluate(model, loader):

    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []  # Store probabilities for ROC curve
    total_loss = 0
    
    with torch.no_grad():
        for batch in tqdm(loader, desc="Evaluating"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()
            
            # Get predictions and probabilities
            logits = outputs.logits
            probs = F.softmax(logits, dim=1)
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())
            all_probs.extend(probs[:, 1].cpu().tolist())  # Probability of positive class

    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    
    # Create confusion matrix
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass  # In case of single class

    mcc = matthews_corrcoef(all_labels, all_preds)

    
    return acc, p, r, f, avg_loss, cm, auc_score, all_labels, all_probs, mcc

def objective(trial):

    # Sample hyperparameters from the trial - adjusted ranges for BERT
    weight_decay = trial.suggest_float("weight_decay", 0.0, 0.3)
    attn_dropout = trial.suggest_float("attention_dropout", 0.0, 0.3)
    hidden_dropout = trial.suggest_float("hidden_dropout", 0.0, 0.3)
    classifier_dropout = trial.suggest_float("classifier_dropout", 0.0, 0.5)  
    
    # Additional BERT-specific hyperparameters
    layer_norm_eps = trial.suggest_float("layer_norm_eps", 1e-12, 1e-6, log=True)

    # Load model config with dropout overrides
    config = AutoConfig.from_pretrained(
        MODEL_NAME,
        num_labels=NUM_CLASSES,
        attention_probs_dropout_prob=attn_dropout,
        hidden_dropout_prob=hidden_dropout,
        classifier_dropout=classifier_dropout,
        layer_norm_eps=layer_norm_eps,  # BERT-specific parameter
        id2label={0: "negative", 1: "positive"},
        label2id={"negative": 0, "positive": 1}
    )

    model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, config=config).to(device)

    optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=weight_decay)
    total_steps = len(train_loader) * 2  # 2 epochs for quick hyperparameter search
    scheduler = get_linear_schedule_with_warmup(optimizer, 0, total_steps)

    best_acc = 0
    for epoch in range(2):  # Reduced epochs for hyperparameter search
        print(f"\n Trial {trial.number + 1}, Epoch {epoch + 1}")
        train_loss, _, train_acc, train_p, train_r, train_f, _, _, _, _, _ = train_epoch(model, train_loader, optimizer, scheduler)
        val_acc, val_p, val_r, val_f, val_loss, _, _, _, _, _ = evaluate(model, val_loader)

        print(f"Train - Loss: {train_loss:.4f} | Acc: {train_acc:.4f} | P: {train_p:.4f} | R: {train_r:.4f} | F1: {train_f:.4f}")
        print(f"Val   - Loss: {val_loss:.4f} | Acc: {val_acc:.4f} | P: {val_p:.4f} | R: {val_r:.4f} | F1: {val_f:.4f}")

        best_acc = max(best_acc, val_acc)

    return best_acc



study = optuna.create_study(direction="maximize")

study.optimize(objective, n_trials=10)


print("\n Best trial results:")
print(f"  F1 Score: {study.best_value:.4f}")
print("  Best hyperparameters:")
for key, value in study.best_trial.params.items():
    print(f"    {key}: {value}")
```

```python

!pip install scikit-learn

!pip install transformers

!pip install emoji contractions

!pip install datasets



import pandas as pd

import numpy as np

import torch

import torch.nn.functional as F

import torch.optim as optim

from torch.utils.data import Dataset, DataLoader


from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_linear_schedule_with_warmup


from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay

from sklearn.metrics import roc_curve, auc, roc_auc_score

import matplotlib.pyplot as plt

import seaborn as sns

from tqdm import tqdm

from datasets import load_dataset

import re

import contractions

import emoji

import string

import random

from torch.utils.data import Dataset, DataLoader, TensorDataset, SequentialSampler

from sklearn.metrics import matthews_corrcoef






TRAIN_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/train_dataset.csv"

VAL_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/val_dataset.csv"

TEST_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/test_dataset.csv"

SUBMISSION_PATH = "/kaggle/working/submission.csv"


MODEL_NAME = "bert-base-uncased"  
BATCH_SIZE = 96  
EPOCHS = 2
LEARNING_RATE = 4e-5  
MAX_LEN = 40
RANDOM_SEED = 42
NUM_CLASSES = 2


ATTENTION_DROPOUT = 0.15      # Standard BERT attention dropout
HIDDEN_DROPOUT = 0.15         # Standard BERT hidden dropout
CLASSIFIER_DROPOUT = 0.15     # BERT classifier dropout
WEIGHT_DECAY = 0.065          # Standard weight decay for BERT

torch.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)


def reduce_repeated_letters(text):

    return re.sub(r"(.)\1{2,}", r"\1\1", text)


def clean_text(text):

    text = str(text).strip()
    
    # Only expand contractions and reduce repeated letters
    text = contractions.fix(text)
    text = reduce_repeated_letters(text)
    
    # Clean up extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text




train_df = pd.read_csv(TRAIN_PATH)

val_df = pd.read_csv(VAL_PATH)

test_df = pd.read_csv(TEST_PATH)


train_df["Text"] = train_df["Text"].apply(clean_text)

val_df["Text"] = val_df["Text"].apply(clean_text)

test_df["Text"] = test_df["Text"].apply(clean_text)



import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np


def analyze_text_lengths(train_df, val_df, test_df, tokenizer):

    
    # Function to get token length for a text
    def get_token_length(text):
        tokens = tokenizer.encode(text, add_special_tokens=True)
        return len(tokens)
    
    # Calculate token lengths for all datasets
    print("Calculating token lengths...")
    train_lengths = [get_token_length(text) for text in train_df["Text"]]
    val_lengths = [get_token_length(text) for text in val_df["Text"]]
    test_lengths = [get_token_length(text) for text in test_df["Text"]]
    
    # Combine all lengths
    all_lengths = train_lengths + val_lengths + test_lengths
    
    # Calculate statistics
    stats = {
        'min': np.min(all_lengths),
        'max': np.max(all_lengths),
        'mean': np.mean(all_lengths),
        'median': np.median(all_lengths),
        'std': np.std(all_lengths),
        'q25': np.percentile(all_lengths, 25),
        'q75': np.percentile(all_lengths, 75),
        'q90': np.percentile(all_lengths, 90),
        'q95': np.percentile(all_lengths, 95),
        'q99': np.percentile(all_lengths, 99)
    }
    
    print("\n TOKEN LENGTH STATISTICS:")
    print(f"Min length: {stats['min']}")
    print(f"Max length: {stats['max']}")
    print(f"Mean length: {stats['mean']:.2f}")
    print(f"Median length: {stats['median']:.2f}")
    print(f"Std deviation: {stats['std']:.2f}")
    print(f"25th percentile: {stats['q25']:.2f}")
    print(f"75th percentile: {stats['q75']:.2f}")
    print(f"90th percentile: {stats['q90']:.2f}")
    print(f"95th percentile: {stats['q95']:.2f}")
    print(f"99th percentile: {stats['q99']:.2f}")
    
    # Calculate coverage for different MAX_LEN values
    max_len_options = [32, 64, 80, 96, 128, 160, 192, 256, 320, 384, 512]
    coverage_data = []
    
    print("\n COVERAGE ANALYSIS FOR DIFFERENT MAX_LEN VALUES:")
    print("MAX_LEN | Coverage | Truncated | Avg Padding")
    print("-" * 45)
    
    for max_len in max_len_options:
        covered = sum(1 for length in all_lengths if length <= max_len)
        coverage_pct = (covered / len(all_lengths)) * 100
        truncated = len(all_lengths) - covered
        avg_padding = max_len - np.mean([min(length, max_len) for length in all_lengths])
        
        coverage_data.append({
            'max_len': max_len,
            'coverage': coverage_pct,
            'truncated': truncated,
            'avg_padding': avg_padding
        })
        
        print(f"{max_len:7d} | {coverage_pct:7.2f}% | {truncated:9d} | {avg_padding:11.2f}")
    
    # Visualization
    plt.figure(figsize=(20, 12))
    
    # 1. Distribution histogram
    plt.subplot(2, 3, 1)
    plt.hist(all_lengths, bins=50, alpha=0.7, edgecolor='black')
    plt.axvline(stats['mean'], color='red', linestyle='--', label=f'Mean: {stats["mean"]:.1f}')
    plt.axvline(stats['median'], color='green', linestyle='--', label=f'Median: {stats["median"]:.1f}')
    plt.axvline(stats['q95'], color='orange', linestyle='--', label=f'95th percentile: {stats["q95"]:.1f}')
    plt.xlabel('Token Length')
    plt.ylabel('Frequency')
    plt.title('Distribution of Token Lengths')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Box plot by dataset
    plt.subplot(2, 3, 2)
    data_for_box = [train_lengths, val_lengths, test_lengths]
    labels = ['Train', 'Validation', 'Test']
    plt.boxplot(data_for_box, labels=labels)
    plt.ylabel('Token Length')
    plt.title('Token Length Distribution by Dataset')
    plt.grid(True, alpha=0.3)
    
    # 3. Cumulative distribution
    plt.subplot(2, 3, 3)
    sorted_lengths = np.sort(all_lengths)
    percentiles = np.arange(1, len(sorted_lengths) + 1) / len(sorted_lengths) * 100
    plt.plot(sorted_lengths, percentiles)
    plt.xlabel('Token Length')
    plt.ylabel('Cumulative Percentage')
    plt.title('Cumulative Distribution of Token Lengths')
    plt.grid(True, alpha=0.3)
    
    # Add common MAX_LEN markers
    for max_len in [64, 80, 128, 256]:
        coverage = (np.searchsorted(sorted_lengths, max_len) / len(sorted_lengths)) * 100
        plt.axvline(max_len, color='red', linestyle='--', alpha=0.5)
        plt.text(max_len, coverage + 2, f'{max_len}', rotation=90, ha='center')
    
    # 4. Coverage vs MAX_LEN
    plt.subplot(2, 3, 4)
    coverages = [item['coverage'] for item in coverage_data]
    max_lens = [item['max_len'] for item in coverage_data]
    plt.plot(max_lens, coverages, marker='o', linewidth=2, markersize=8)
    plt.xlabel('MAX_LEN')
    plt.ylabel('Coverage (%)')
    plt.title('Coverage vs MAX_LEN')
    plt.grid(True, alpha=0.3)
    
    # Highlight good options
    for i, (max_len, coverage) in enumerate(zip(max_lens, coverages)):
        if coverage >= 95:
            plt.annotate(f'{max_len}\n({coverage:.1f}%)', 
                        (max_len, coverage), 
                        textcoords="offset points", 
                        xytext=(0,10), 
                        ha='center', 
                        fontsize=9,
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    # 5. Efficiency analysis (Coverage vs Padding)
    plt.subplot(2, 3, 5)
    avg_paddings = [item['avg_padding'] for item in coverage_data]
    plt.scatter(avg_paddings, coverages, s=100, alpha=0.7)
    
    for i, max_len in enumerate(max_lens):
        plt.annotate(str(max_len), (avg_paddings[i], coverages[i]), 
                    textcoords="offset points", xytext=(5,5), ha='left')
    
    plt.xlabel('Average Padding')
    plt.ylabel('Coverage (%)')
    plt.title('Efficiency: Coverage vs Average Padding')
    plt.grid(True, alpha=0.3)
    
    # 6. Length distribution by class (if labels available)
    plt.subplot(2, 3, 6)
    if 'Label' in train_df.columns:
        neg_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 0]
        pos_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 1]
        
        plt.hist(neg_lengths, bins=30, alpha=0.7, label='Negative', color='red')
        plt.hist(pos_lengths, bins=30, alpha=0.7, label='Positive', color='blue')
        plt.xlabel('Token Length')
        plt.ylabel('Frequency')
        plt.title('Token Length Distribution by Class')
        plt.legend()
        plt.grid(True, alpha=0.3)
    else:
        plt.text(0.5, 0.5, 'No labels available\nfor class analysis', 
                ha='center', va='center', transform=plt.gca().transAxes,
                fontsize=12, bbox=dict(boxstyle="round", facecolor='lightgray'))
        plt.title('Class Analysis Not Available')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/max_len_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Recommendations
    print("\nRECOMMENDATIONS:")
    print("=" * 40)
    
    # Find the most efficient MAX_LEN values
    efficient_options = []
    for item in coverage_data:
        if item['coverage'] >= 90:  # At least 90% coverage
            efficiency = item['coverage'] / (item['max_len'] / stats['mean'])  # Coverage per relative length
            efficient_options.append((item['max_len'], item['coverage'], efficiency))
    
    efficient_options.sort(key=lambda x: x[2], reverse=True)  # Sort by efficiency
    
    print("Most efficient MAX_LEN options (>=90% coverage):")
    for max_len, coverage, efficiency in efficient_options[:3]:
        print(f"  MAX_LEN = {max_len}: {coverage:.1f}% coverage (efficiency: {efficiency:.2f})")
    
    # Specific recommendations based on percentiles
    print(f"\nBased on percentiles:")
    print(f"  Conservative (99% coverage): MAX_LEN = {int(stats['q99']) + 5}")
    print(f"  Balanced (95% coverage): MAX_LEN = {int(stats['q95']) + 5}")
    print(f"  Aggressive (90% coverage): MAX_LEN = {int(stats['q90']) + 5}")
    print(f"  Very aggressive (75% coverage): MAX_LEN = {int(stats['q75']) + 5}")
    
    # Return the statistics for further use
    return stats, coverage_data


print(" Starting EDA for MAX_LEN optimization...")

stats, coverage_data = analyze_text_lengths(train_df, val_df, test_df, tokenizer)


optimal_max_len = int(stats['q95']) + 5  # 95th percentile + buffer

print(f"\nSUGGESTED MAX_LEN: {optimal_max_len}")



class TweetDataset(Dataset):

    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long)
        }


train_dataset = TweetDataset(train_df["Text"].tolist(), train_df["Label"].tolist(), tokenizer, MAX_LEN)

val_dataset = TweetDataset(val_df["Text"].tolist(), val_df["Label"].tolist(), tokenizer, MAX_LEN)


train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)

val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)



from transformers import AutoConfig


config = AutoConfig.from_pretrained(MODEL_NAME)


config.attention_probs_dropout_prob = ATTENTION_DROPOUT  # Attention dropout

config.hidden_dropout_prob = HIDDEN_DROPOUT              # Hidden layer dropout


if hasattr(config, 'classifier_dropout_prob'):

    config.classifier_dropout_prob = CLASSIFIER_DROPOUT
elif hasattr(config, 'classifier_dropout'):

    config.classifier_dropout = CLASSIFIER_DROPOUT
    

config.num_labels = NUM_CLASSES

config.id2label = {0: "negative", 1: "positive"}

config.label2id = {"negative": 0, "positive": 1}


model = AutoModelForSequenceClassification.from_pretrained(

    MODEL_NAME,
    config=config
).to(device)




def create_optimizer_with_weight_decay(model, learning_rate, weight_decay):

    # Separate parameters that should and shouldn't have weight decay
    decay_parameters = []
    no_decay_parameters = []
    
    for name, param in model.named_parameters():
        if not param.requires_grad:
            continue
            
        # Don't apply weight decay to bias terms and layer norm parameters
        if any(nd in name for nd in ["bias", "LayerNorm.weight", "layer_norm.weight"]):
            no_decay_parameters.append(param)
        else:
            decay_parameters.append(param)
    
    # Create parameter groups
    optimizer_grouped_parameters = [
        {
            "params": decay_parameters,
            "weight_decay": weight_decay,
            "lr": learning_rate
        },
        {
            "params": no_decay_parameters,
            "weight_decay": 0.0,
            "lr": learning_rate
        }
    ]
    

    return optim.AdamW(optimizer_grouped_parameters)


optimizer = create_optimizer_with_weight_decay(model, LEARNING_RATE, WEIGHT_DECAY)



total_steps = len(train_loader) * EPOCHS

warmup_steps = int(0.1 * total_steps)  # 10% warmup

scheduler = get_linear_schedule_with_warmup(optimizer, warmup_steps, total_steps)





def train_epoch(model, loader):

    model.train()
    total_loss = 0
    batch_losses = []
    all_preds = []
    all_labels = []
    all_probs = []
    
    for batch in tqdm(loader, desc="Training"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()
        batch_losses.append(loss.item())

        # Get predictions and track metrics during training
        logits = outputs.logits
        probs = F.softmax(logits, dim=1)
        preds = torch.argmax(logits, dim=1)
        
        # Collect for metrics calculation
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())
        all_probs.extend(probs[:, 1].cpu().tolist())

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()
        
    # Calculate metrics
    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass
    
    mcc = matthews_corrcoef(all_labels, all_preds)

    return avg_loss, batch_losses, acc, p, r, f, cm, auc_score, all_labels, all_probs, mcc


def evaluate(model, loader):

    model.eval()  
    all_preds = []
    all_labels = []
    all_probs = []
    total_loss = 0
    
    with torch.no_grad():
        for batch in tqdm(loader, desc="Evaluating"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()
            
            # Get predictions and probabilities
            logits = outputs.logits
            probs = F.softmax(logits, dim=1)
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())
            all_probs.extend(probs[:, 1].cpu().tolist())

    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass

    mcc = matthews_corrcoef(all_labels, all_preds)
    
    return acc, p, r, f, avg_loss, cm, auc_score, all_labels, all_probs, mcc


def plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, 
                 model_name, batch_losses=None):
                 
    """Plot training and validation metrics"""
    plt.figure(figsize=(15, 10))
    
    # Plot losses
    plt.subplot(2, 2, 1)
    plt.plot(train_losses, label="Train Loss", marker='o')
    plt.plot(val_losses, label="Validation Loss", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid(True)
    
    # Plot accuracy
    plt.subplot(2, 2, 2)
    plt.plot(train_accs, label="Train Accuracy", marker='o')
    plt.plot(val_accs, label="Validation Accuracy", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Training and Validation Accuracy")
    plt.legend()
    plt.grid(True)
    
    # Plot F1 score
    plt.subplot(2, 2, 3)
    plt.plot(train_f1s, label="Train F1", marker='o')
    plt.plot(val_f1s, label="Validation F1", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("F1 Score")
    plt.title("Training and Validation F1 Score")
    plt.legend()
    plt.grid(True)
    
    # Plot batch-level losses if available
    if batch_losses:
        plt.subplot(2, 2, 4)
        all_batches = []
        for epoch, losses in enumerate(batch_losses):
            batches = list(range(epoch * len(losses), (epoch + 1) * len(losses)))
            all_batches.extend(batches)
            plt.plot(batches, losses, alpha=0.5, label=f"Epoch {epoch+1}")
        
        plt.xlabel("Batch")
        plt.ylabel("Loss")
        plt.title("Batch-level Loss")
        plt.legend()
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig(f"/kaggle/working/{model_name}_learning_curves.png")
    plt.show()

def plot_confusion_matrix(cm, model_name, set_name="Validation"):

    """Plot confusion matrix"""
    plt.figure(figsize=(8, 6))
    display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
    display.plot(cmap=plt.cm.Blues, values_format=".0f")
    plt.title(f"{set_name} Confusion Matrix")
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_confusion_matrix.png")
    plt.show()

def plot_roc_curve(y_true, y_probs, model_name, set_name="Validation"):

    """Plot ROC curve"""
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{set_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_roc_curve.png")
    plt.show()
    
    return roc_auc


best_f1 = 0

model_path = f"/kaggle/working/best_{MODEL_NAME.replace('-', '_')}.pth"


train_losses = []
val_losses = []
train_accs = []
val_accs = []
train_f1s = []
val_f1s = []
train_aucs = []
val_aucs = []
batch_level_losses = []

for epoch in range(EPOCHS):

    print(f"\nEpoch {epoch + 1}/{EPOCHS}")
    
    # Train and collect metrics directly during training
    train_loss, batch_losses, train_acc, train_p, train_r, train_f1, train_cm, train_auc, train_labels, train_probs, train_mcc = train_epoch(model, train_loader)
    batch_level_losses.append(batch_losses)
    
    # Evaluate only on validation set
    val_acc, val_p, val_r, val_f1, val_loss, val_cm, val_auc, val_labels, val_probs, val_mcc = evaluate(model, val_loader)

    # Store metrics
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    train_f1s.append(train_f1)
    val_f1s.append(val_f1)
    
    # Add AUC scores if available
    if train_auc is not None:
        train_aucs.append(train_auc)
    if val_auc is not None:
        val_aucs.append(val_auc)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Train  - Acc: {train_acc:.4f}, Precision: {train_p:.4f}, Recall: {train_r:.4f}, F1: {train_f1:.4f}")
    if train_auc is not None:
        print(f"         AUC: {train_auc:.4f}")
    
    print(f"Val    - Acc: {val_acc:.4f}, Precision: {val_p:.4f}, Recall: {val_r:.4f}, F1: {val_f1:.4f}, Loss: {val_loss:.4f}")
    if val_auc is not None:
        print(f"         AUC: {val_auc:.4f}")

    # Save best model
    if val_f1 > best_f1:
        best_f1 = val_f1
        torch.save(model.state_dict(), model_path)
        print(" Best model saved!")
        
        # Plot confusion matrix and ROC curve for best model
        plot_confusion_matrix(val_cm, MODEL_NAME)
        if val_auc is not None:
            plot_roc_curve(val_labels, val_probs, MODEL_NAME)


plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, MODEL_NAME, batch_level_losses)



model.load_state_dict(torch.load(model_path))

model.eval()


final_val_acc, final_val_p, final_val_r, final_val_f1, final_val_loss, final_val_cm, final_val_auc, final_val_labels, final_val_probs, final_val_mcc = evaluate(model, val_loader)

print("\n===== FINAL MODEL EVALUATION =====")

print(f"Val    - Acc: {final_val_acc:.4f}, Precision: {final_val_p:.4f}, Recall: {final_val_r:.4f}, F1: {final_val_f1:.4f}, MCC: {final_val_mcc:.4f}, Loss: {final_val_loss:.4f}")

if final_val_auc is not None:

    print(f"         AUC: {final_val_auc:.4f}")


plot_confusion_matrix(final_val_cm, MODEL_NAME, "Validation")

if final_val_auc is not None:

    plot_roc_curve(final_val_labels, final_val_probs, MODEL_NAME, "Final Validation")


class TestDataset(Dataset):

    def __init__(self, texts, tokenizer, max_len):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0)
        }

test_dataset = TestDataset(test_df["Text"].tolist(), tokenizer, MAX_LEN)

test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)


test_preds = []

with torch.no_grad():

    for batch in tqdm(test_loader, desc="Predicting"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1)
        
        test_preds.extend(preds.cpu().tolist())


id_col = "ID" if "ID" in test_df.columns else "Id"

submission = pd.DataFrame({id_col: test_df[id_col], "Label": test_preds})

submission.to_csv(SUBMISSION_PATH, index=False)

print(f"Submission saved to: {SUBMISSION_PATH}")
```

```python
!pip install scikit-learn

!pip install transformers

!pip install emoji contractions

!pip install datasets


import pandas as pd

import numpy as np

import torch

import torch.nn.functional as F

import torch.optim as optim

from torch.utils.data import Dataset, DataLoader


from transformers import BertTokenizer, BertForSequenceClassification

from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts


from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay

from sklearn.metrics import roc_curve, auc, roc_auc_score

import matplotlib.pyplot as plt

import seaborn as sns

from tqdm import tqdm

from datasets import load_dataset

import re

import contractions

import emoji

import string

import random

from torch.utils.data import Dataset, DataLoader, TensorDataset, SequentialSampler

from sklearn.metrics import matthews_corrcoef





TRAIN_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/train_dataset.csv"

VAL_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/val_dataset.csv"

TEST_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/test_dataset.csv"

SUBMISSION_PATH = "/kaggle/working/submission.csv"


MODEL_NAME = "bert-base-uncased" 

BATCH_SIZE = 16

EPOCHS = 3

LEARNING_RATE = 4e-5

MAX_LEN = 40

RANDOM_SEED = 42

NUM_CLASSES = 2


ATTENTION_DROPOUT = 0.07     # Dropout for attention weights

HIDDEN_DROPOUT = 0.15         # Dropout for hidden states

CLASSIFIER_DROPOUT = 0.01     # Dropout for classifier layer (higher for final layer)

WEIGHT_DECAY = 0.5         # L2 regularization strength


T_0 = 50                    # Number of iterations for the first restart

T_MULT = 2                  # A factor increases T_i after a restart

ETA_MIN = 1e-7              # Minimum learning rate

LAST_EPOCH = -1             # The index of last epoch


torch.manual_seed(RANDOM_SEED)

np.random.seed(RANDOM_SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")



tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)



def reduce_repeated_letters(text):

    return re.sub(r"(.)\1{2,}", r"\1\1", text)

def clean_text(text):

    text = str(text).strip()
    
    # Expand contractions and reduce repeated letters
    text = contractions.fix(text)
    text = reduce_repeated_letters(text)
    
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text



train_df = pd.read_csv(TRAIN_PATH)

val_df = pd.read_csv(VAL_PATH)

test_df = pd.read_csv(TEST_PATH)


train_df["Text"] = train_df["Text"].apply(clean_text)

val_df["Text"] = val_df["Text"].apply(clean_text)

test_df["Text"] = test_df["Text"].apply(clean_text)



import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np


def analyze_text_lengths(train_df, val_df, test_df, tokenizer):

    """
    Analyze text lengths to determine optimal MAX_LEN
    """
    print(" ANALYZING TEXT LENGTHS FOR MAX_LEN OPTIMIZATION")
    print("=" * 60)
    
    # Function to get token length for a text
    def get_token_length(text):
        tokens = tokenizer.encode(text, add_special_tokens=True)
        return len(tokens)
    
    # Calculate token lengths for all datasets
    print(" Calculating token lengths...")
    train_lengths = [get_token_length(text) for text in train_df["Text"]]
    val_lengths = [get_token_length(text) for text in val_df["Text"]]
    test_lengths = [get_token_length(text) for text in test_df["Text"]]
    
    # Combine all lengths
    all_lengths = train_lengths + val_lengths + test_lengths
    
    # Calculate statistics
    stats = {
        'min': np.min(all_lengths),
        'max': np.max(all_lengths),
        'mean': np.mean(all_lengths),
        'median': np.median(all_lengths),
        'std': np.std(all_lengths),
        'q25': np.percentile(all_lengths, 25),
        'q75': np.percentile(all_lengths, 75),
        'q90': np.percentile(all_lengths, 90),
        'q95': np.percentile(all_lengths, 95),
        'q99': np.percentile(all_lengths, 99)
    }
    
    print("\n TOKEN LENGTH STATISTICS:")
    print(f"Min length: {stats['min']}")
    print(f"Max length: {stats['max']}")
    print(f"Mean length: {stats['mean']:.2f}")
    print(f"Median length: {stats['median']:.2f}")
    print(f"Std deviation: {stats['std']:.2f}")
    print(f"25th percentile: {stats['q25']:.2f}")
    print(f"75th percentile: {stats['q75']:.2f}")
    print(f"90th percentile: {stats['q90']:.2f}")
    print(f"95th percentile: {stats['q95']:.2f}")
    print(f"99th percentile: {stats['q99']:.2f}")
    
    # Calculate coverage for different MAX_LEN values
    max_len_options = [32, 64, 80, 96, 128, 160, 192, 256, 320, 384, 512]
    coverage_data = []
    
    print("\n COVERAGE ANALYSIS FOR DIFFERENT MAX_LEN VALUES:")
    print("MAX_LEN | Coverage | Truncated | Avg Padding")
    print("-" * 45)
    
    for max_len in max_len_options:
        covered = sum(1 for length in all_lengths if length <= max_len)
        coverage_pct = (covered / len(all_lengths)) * 100
        truncated = len(all_lengths) - covered
        avg_padding = max_len - np.mean([min(length, max_len) for length in all_lengths])
        
        coverage_data.append({
            'max_len': max_len,
            'coverage': coverage_pct,
            'truncated': truncated,
            'avg_padding': avg_padding
        })
        
        print(f"{max_len:7d} | {coverage_pct:7.2f}% | {truncated:9d} | {avg_padding:11.2f}")
    
    # Visualization
    plt.figure(figsize=(20, 12))
    
    # 1. Distribution histogram
    plt.subplot(2, 3, 1)
    plt.hist(all_lengths, bins=50, alpha=0.7, edgecolor='black')
    plt.axvline(stats['mean'], color='red', linestyle='--', label=f'Mean: {stats["mean"]:.1f}')
    plt.axvline(stats['median'], color='green', linestyle='--', label=f'Median: {stats["median"]:.1f}')
    plt.axvline(stats['q95'], color='orange', linestyle='--', label=f'95th percentile: {stats["q95"]:.1f}')
    plt.xlabel('Token Length')
    plt.ylabel('Frequency')
    plt.title('Distribution of Token Lengths')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Box plot by dataset
    plt.subplot(2, 3, 2)
    data_for_box = [train_lengths, val_lengths, test_lengths]
    labels = ['Train', 'Validation', 'Test']
    plt.boxplot(data_for_box, labels=labels)
    plt.ylabel('Token Length')
    plt.title('Token Length Distribution by Dataset')
    plt.grid(True, alpha=0.3)
    
    # 3. Cumulative distribution
    plt.subplot(2, 3, 3)
    sorted_lengths = np.sort(all_lengths)
    percentiles = np.arange(1, len(sorted_lengths) + 1) / len(sorted_lengths) * 100
    plt.plot(sorted_lengths, percentiles)
    plt.xlabel('Token Length')
    plt.ylabel('Cumulative Percentage')
    plt.title('Cumulative Distribution of Token Lengths')
    plt.grid(True, alpha=0.3)
    
    # Add common MAX_LEN markers
    for max_len in [64, 80, 128, 256]:
        coverage = (np.searchsorted(sorted_lengths, max_len) / len(sorted_lengths)) * 100
        plt.axvline(max_len, color='red', linestyle='--', alpha=0.5)
        plt.text(max_len, coverage + 2, f'{max_len}', rotation=90, ha='center')
    
    # 4. Coverage vs MAX_LEN
    plt.subplot(2, 3, 4)
    coverages = [item['coverage'] for item in coverage_data]
    max_lens = [item['max_len'] for item in coverage_data]
    plt.plot(max_lens, coverages, marker='o', linewidth=2, markersize=8)
    plt.xlabel('MAX_LEN')
    plt.ylabel('Coverage (%)')
    plt.title('Coverage vs MAX_LEN')
    plt.grid(True, alpha=0.3)
    
    # Highlight good options
    for i, (max_len, coverage) in enumerate(zip(max_lens, coverages)):
        if coverage >= 95:
            plt.annotate(f'{max_len}\n({coverage:.1f}%)', 
                        (max_len, coverage), 
                        textcoords="offset points", 
                        xytext=(0,10), 
                        ha='center', 
                        fontsize=9,
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    # 5. Efficiency analysis (Coverage vs Padding)
    plt.subplot(2, 3, 5)
    avg_paddings = [item['avg_padding'] for item in coverage_data]
    plt.scatter(avg_paddings, coverages, s=100, alpha=0.7)
    
    for i, max_len in enumerate(max_lens):
        plt.annotate(str(max_len), (avg_paddings[i], coverages[i]), 
                    textcoords="offset points", xytext=(5,5), ha='left')
    
    plt.xlabel('Average Padding')
    plt.ylabel('Coverage (%)')
    plt.title('Efficiency: Coverage vs Average Padding')
    plt.grid(True, alpha=0.3)
    
    # 6. Length distribution by class (if labels available)
    plt.subplot(2, 3, 6)
    if 'Label' in train_df.columns:
        neg_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 0]
        pos_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 1]
        
        plt.hist(neg_lengths, bins=30, alpha=0.7, label='Negative', color='red')
        plt.hist(pos_lengths, bins=30, alpha=0.7, label='Positive', color='blue')
        plt.xlabel('Token Length')
        plt.ylabel('Frequency')
        plt.title('Token Length Distribution by Class')
        plt.legend()
        plt.grid(True, alpha=0.3)
    else:
        plt.text(0.5, 0.5, 'No labels available\nfor class analysis', 
                ha='center', va='center', transform=plt.gca().transAxes,
                fontsize=12, bbox=dict(boxstyle="round", facecolor='lightgray'))
        plt.title('Class Analysis Not Available')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/max_len_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Recommendations
    print("\n RECOMMENDATIONS:")
    print("=" * 40)
    
    # Find the most efficient MAX_LEN values
    efficient_options = []
    for item in coverage_data:
        if item['coverage'] >= 90:  # At least 90% coverage
            efficiency = item['coverage'] / (item['max_len'] / stats['mean'])  # Coverage per relative length
            efficient_options.append((item['max_len'], item['coverage'], efficiency))
    
    efficient_options.sort(key=lambda x: x[2], reverse=True)  # Sort by efficiency
    
    print("Most efficient MAX_LEN options (>=90% coverage):")
    for max_len, coverage, efficiency in efficient_options[:3]:
        print(f"  MAX_LEN = {max_len}: {coverage:.1f}% coverage (efficiency: {efficiency:.2f})")
    
    # Specific recommendations based on percentiles
    print(f"\nBased on percentiles:")
    print(f"  Conservative (99% coverage): MAX_LEN = {int(stats['q99']) + 5}")
    print(f"  Balanced (95% coverage): MAX_LEN = {int(stats['q95']) + 5}")
    print(f"  Aggressive (90% coverage): MAX_LEN = {int(stats['q90']) + 5}")
    print(f"  Very aggressive (75% coverage): MAX_LEN = {int(stats['q75']) + 5}")
    
    # Return the statistics for further use
    return stats, coverage_data

print("Starting EDA for MAX_LEN optimization...")
stats, coverage_data = analyze_text_lengths(train_df, val_df, test_df, tokenizer)

optimal_max_len = int(stats['q95']) + 5  # 95th percentile + buffer
print(f"\n SUGGESTED MAX_LEN: {optimal_max_len}")


class TweetDataset(Dataset):

    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long)
        }


train_dataset = TweetDataset(train_df["Text"].tolist(), train_df["Label"].tolist(), tokenizer, MAX_LEN)

val_dataset = TweetDataset(val_df["Text"].tolist(), val_df["Label"].tolist(), tokenizer, MAX_LEN)


train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)

val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)



from transformers import BertConfig


config = BertConfig.from_pretrained(MODEL_NAME)


config.attention_probs_dropout_prob = ATTENTION_DROPOUT  # Attention dropout

config.hidden_dropout_prob = HIDDEN_DROPOUT              # Hidden layer dropout



if hasattr(config, 'classifier_dropout'):

    config.classifier_dropout = CLASSIFIER_DROPOUT
    


config.num_labels = NUM_CLASSES

config.id2label = {0: "negative", 1: "positive"}

config.label2id = {"negative": 0, "positive": 1}



model = BertForSequenceClassification.from_pretrained(

    MODEL_NAME,
    config=config
).to(device)


def create_optimizer_with_weight_decay(model, learning_rate, weight_decay):

    """
    Create optimizer with proper weight decay configuration.
    Apply weight decay to weights but not to biases and layer norm parameters.
    """
    # Separate parameters that should and shouldn't have weight decay
    decay_parameters = []
    no_decay_parameters = []
    
    for name, param in model.named_parameters():
        if not param.requires_grad:
            continue
            
        # Don't apply weight decay to bias terms and layer norm parameters
        if any(nd in name for nd in ["bias", "LayerNorm.weight", "layer_norm.weight"]):
            no_decay_parameters.append(param)
        else:
            decay_parameters.append(param)
    
    # Create parameter groups
    optimizer_grouped_parameters = [
        {
            "params": decay_parameters,
            "weight_decay": weight_decay,
            "lr": learning_rate
        },
        {
            "params": no_decay_parameters,
            "weight_decay": 0.0,
            "lr": learning_rate
        }
    ]
    
    
    return optim.AdamW(optimizer_grouped_parameters)

optimizer = create_optimizer_with_weight_decay(model, LEARNING_RATE, WEIGHT_DECAY)



total_steps = len(train_loader) * EPOCHS


scheduler = CosineAnnealingWarmRestarts(

    optimizer,
    T_0=T_0,              # Number of iterations for the first restart
    T_mult=T_MULT,        # A factor increases T_i after a restart
    eta_min=ETA_MIN,      # Minimum learning rate
    last_epoch=LAST_EPOCH # The index of last epoch
)


def train_epoch(model, loader, optimizer, scheduler):

    model.train()
    total_loss = 0
    batch_losses = []
    all_preds = []
    all_labels = []
    all_probs = []
    learning_rates = []  # Track learning rate changes
    
    for batch_idx, batch in enumerate(tqdm(loader, desc="Training")):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()
        batch_losses.append(loss.item())

        # Get predictions and track metrics during training
        logits = outputs.logits
        probs = F.softmax(logits, dim=1)
        preds = torch.argmax(logits, dim=1)
        
        # Collect for metrics calculation
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())
        all_probs.extend(probs[:, 1].cpu().tolist())

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        
        # Step the scheduler after each batch for CosineAnnealingWarmRestarts
        scheduler.step()
        
        # Track learning rate
        current_lr = optimizer.param_groups[0]['lr']
        learning_rates.append(current_lr)
        
        optimizer.zero_grad()
        
    # Calculate metrics
    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass
    
    mcc = matthews_corrcoef(all_labels, all_preds)

    return avg_loss, batch_losses, acc, p, r, f, cm, auc_score, all_labels, all_probs, mcc, learning_rates


def evaluate(model, loader):

    model.eval()  # Important: Set to eval mode to disable dropout
    all_preds = []
    all_labels = []
    all_probs = []
    total_loss = 0
    
    with torch.no_grad():
        for batch in tqdm(loader, desc="Evaluating"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()
            
            # Get predictions and probabilities
            logits = outputs.logits
            probs = F.softmax(logits, dim=1)
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())
            all_probs.extend(probs[:, 1].cpu().tolist())

    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass

    mcc = matthews_corrcoef(all_labels, all_preds)
    
    return acc, p, r, f, avg_loss, cm, auc_score, all_labels, all_probs, mcc



def plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, 
                 model_name, batch_losses=None):
                 
    """Plot training and validation metrics"""
    plt.figure(figsize=(15, 10))
    
    # Plot losses
    plt.subplot(2, 2, 1)
    plt.plot(train_losses, label="Train Loss", marker='o')
    plt.plot(val_losses, label="Validation Loss", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid(True)
    
    # Plot accuracy
    plt.subplot(2, 2, 2)
    plt.plot(train_accs, label="Train Accuracy", marker='o')
    plt.plot(val_accs, label="Validation Accuracy", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Training and Validation Accuracy")
    plt.legend()
    plt.grid(True)
    
    # Plot F1 score
    plt.subplot(2, 2, 3)
    plt.plot(train_f1s, label="Train F1", marker='o')
    plt.plot(val_f1s, label="Validation F1", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("F1 Score")
    plt.title("Training and Validation F1 Score")
    plt.legend()
    plt.grid(True)
    
    # Plot batch-level losses if available
    if batch_losses:
        plt.subplot(2, 2, 4)
        all_batches = []
        for epoch, losses in enumerate(batch_losses):
            batches = list(range(epoch * len(losses), (epoch + 1) * len(losses)))
            all_batches.extend(batches)
            plt.plot(batches, losses, alpha=0.5, label=f"Epoch {epoch+1}")
        
        plt.xlabel("Batch")
        plt.ylabel("Loss")
        plt.title("Batch-level Loss")
        plt.legend()
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig(f"/kaggle/working/{model_name}_learning_curves.png")
    plt.show()

def plot_confusion_matrix(cm, model_name, set_name="Validation"):

    """Plot confusion matrix"""
    plt.figure(figsize=(8, 6))
    display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
    display.plot(cmap=plt.cm.Blues, values_format=".0f")
    plt.title(f"{set_name} Confusion Matrix")
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_confusion_matrix.png")
    plt.show()

def plot_roc_curve(y_true, y_probs, model_name, set_name="Validation"):

    """Plot ROC curve"""
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{set_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_roc_curve.png")
    plt.show()
    
    return roc_auc


best_f1 = 0

model_path = f"/kaggle/working/best_{MODEL_NAME.replace('-', '_')}.pth"


train_losses = []

val_losses = []

train_accs = []

val_accs = []

train_f1s = []

val_f1s = []

train_aucs = []

val_aucs = []

batch_level_losses = []


for epoch in range(EPOCHS):

    print(f"\nEpoch {epoch + 1}/{EPOCHS}")
    
    # Train and collect metrics directly during training
    train_loss, batch_losses, train_acc, train_p, train_r, train_f1, train_cm, train_auc, train_labels, train_probs, train_mcc, _ = train_epoch(model, train_loader, optimizer, scheduler)
    batch_level_losses.append(batch_losses)
    
    # Evaluate only on validation set
    val_acc, val_p, val_r, val_f1, val_loss, val_cm, val_auc, val_labels, val_probs, val_mcc = evaluate(model, val_loader)

    # Store metrics
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    train_f1s.append(train_f1)
    val_f1s.append(val_f1)
    
    # Add AUC scores if available
    if train_auc is not None:
        train_aucs.append(train_auc)
    if val_auc is not None:
        val_aucs.append(val_auc)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Train  - Acc: {train_acc:.4f}, Precision: {train_p:.4f}, Recall: {train_r:.4f}, F1: {train_f1:.4f}")
    if train_auc is not None:
        print(f"         AUC: {train_auc:.4f}")
    
    print(f"Val    - Acc: {val_acc:.4f}, Precision: {val_p:.4f}, Recall: {val_r:.4f}, F1: {val_f1:.4f}, Loss: {val_loss:.4f}")
    if val_auc is not None:
        print(f"         AUC: {val_auc:.4f}")

    # Save best model
    if val_f1 > best_f1:
        best_f1 = val_f1
        torch.save(model.state_dict(), model_path)
        print(" Best model saved!")
        
        # Plot confusion matrix and ROC curve for best model
        plot_confusion_matrix(val_cm, MODEL_NAME)
        if val_auc is not None:
            plot_roc_curve(val_labels, val_probs, MODEL_NAME)


plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, MODEL_NAME, batch_level_losses)



model.load_state_dict(torch.load(model_path))

model.eval()


final_val_acc, final_val_p, final_val_r, final_val_f1, final_val_loss, final_val_cm, final_val_auc, final_val_labels, final_val_probs, final_val_mcc = evaluate(model, val_loader)

print("\n===== FINAL MODEL EVALUATION =====")

print(f"Val    - Acc: {final_val_acc:.4f}, Precision: {final_val_p:.4f}, Recall: {final_val_r:.4f}, F1: {final_val_f1:.4f}, MCC: {final_val_mcc:.4f}, Loss: {final_val_loss:.4f}")

if final_val_auc is not None:

    print(f"         AUC: {final_val_auc:.4f}")


plot_confusion_matrix(final_val_cm, MODEL_NAME, "Validation")

if final_val_auc is not None:

    plot_roc_curve(final_val_labels, final_val_probs, MODEL_NAME, "Final Validation")


class TestDataset(Dataset):

    def __init__(self, texts, tokenizer, max_len):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0)
        }

test_dataset = TestDataset(test_df["Text"].tolist(), tokenizer, MAX_LEN)

test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)


test_preds = []

with torch.no_grad():

    for batch in tqdm(test_loader, desc="Predicting"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1)
        
        test_preds.extend(preds.cpu().tolist())

id_col = "ID" if "ID" in test_df.columns else "Id"

submission = pd.DataFrame({id_col: test_df[id_col], "Label": test_preds})

submission.to_csv(SUBMISSION_PATH, index=False)

print(f" Submission saved to: {SUBMISSION_PATH}") 
```

```python
!pip install scikit-learn

!pip install transformers

!pip install emoji contractions

!pip install datasets



import pandas as pd

import numpy as np

import torch

import torch.nn.functional as F

import torch.optim as optim

from torch.utils.data import Dataset, DataLoader


from transformers import AutoTokenizer, AutoModelForSequenceClassification

from torch.optim.lr_scheduler import ReduceLROnPlateau


from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay

from sklearn.metrics import roc_curve, auc, roc_auc_score

import matplotlib.pyplot as plt

import seaborn as sns

from tqdm import tqdm

from datasets import load_dataset

import re

import contractions

import emoji

import string

import random

from torch.utils.data import Dataset, DataLoader, TensorDataset, SequentialSampler

from sklearn.metrics import matthews_corrcoef

from torch.optim.lr_scheduler import ReduceLROnPlateau



TRAIN_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/train_dataset.csv"

VAL_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/val_dataset.csv"

TEST_PATH = "/kaggle/input/ai-2-dl-for-nlp-2025-homework-3/test_dataset.csv"

SUBMISSION_PATH = "/kaggle/working/submission.csv"


MODEL_NAME = "bert-base-uncased"  

BATCH_SIZE = 16  

EPOCHS = 3

LEARNING_RATE = 4e-5  

MAX_LEN = 40

RANDOM_SEED = 42

NUM_CLASSES = 2



ATTENTION_DROPOUT = 0.07      # Standard dropout for BERT attention

HIDDEN_DROPOUT = 0.3         # Standard dropout for BERT hidden states

CLASSIFIER_DROPOUT = 0.001    # Standard dropout for classifier layer

WEIGHT_DECAY = 0.25       # Standard weight decay for BERT


torch.manual_seed(RANDOM_SEED)

np.random.seed(RANDOM_SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")



tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)




def reduce_repeated_letters(text):

    return re.sub(r"(.)\1{2,}", r"\1\1", text)


def clean_text(text):

    text = str(text).strip()
    
    # Only expand contractions and reduce repeated letters
    text = contractions.fix(text)
    text = reduce_repeated_letters(text)
    
    # Clean up extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

train_df = pd.read_csv(TRAIN_PATH)

val_df = pd.read_csv(VAL_PATH)

test_df = pd.read_csv(TEST_PATH)


train_df["Text"] = train_df["Text"].apply(clean_text)

val_df["Text"] = val_df["Text"].apply(clean_text)

test_df["Text"] = test_df["Text"].apply(clean_text)



import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np


def analyze_text_lengths(train_df, val_df, test_df, tokenizer):

    print(" ANALYZING TEXT LENGTHS FOR MAX_LEN OPTIMIZATION")
    print("=" * 60)
    
    # Function to get token length for a text
    def get_token_length(text):
        tokens = tokenizer.encode(text, add_special_tokens=True)
        return len(tokens)
    
    # Calculate token lengths for all datasets
    print(" Calculating token lengths...")
    train_lengths = [get_token_length(text) for text in train_df["Text"]]
    val_lengths = [get_token_length(text) for text in val_df["Text"]]
    test_lengths = [get_token_length(text) for text in test_df["Text"]]
    
    # Combine all lengths
    all_lengths = train_lengths + val_lengths + test_lengths
    
    # Calculate statistics
    stats = {
        'min': np.min(all_lengths),
        'max': np.max(all_lengths),
        'mean': np.mean(all_lengths),
        'median': np.median(all_lengths),
        'std': np.std(all_lengths),
        'q25': np.percentile(all_lengths, 25),
        'q75': np.percentile(all_lengths, 75),
        'q90': np.percentile(all_lengths, 90),
        'q95': np.percentile(all_lengths, 95),
        'q99': np.percentile(all_lengths, 99)
    }
    
    print("\n TOKEN LENGTH STATISTICS:")
    print(f"Min length: {stats['min']}")
    print(f"Max length: {stats['max']}")
    print(f"Mean length: {stats['mean']:.2f}")
    print(f"Median length: {stats['median']:.2f}")
    print(f"Std deviation: {stats['std']:.2f}")
    print(f"25th percentile: {stats['q25']:.2f}")
    print(f"75th percentile: {stats['q75']:.2f}")
    print(f"90th percentile: {stats['q90']:.2f}")
    print(f"95th percentile: {stats['q95']:.2f}")
    print(f"99th percentile: {stats['q99']:.2f}")
    
    # Calculate coverage for different MAX_LEN values
    max_len_options = [32, 64, 80, 96, 128, 160, 192, 256, 320, 384, 512]
    coverage_data = []
    
    print("\n COVERAGE ANALYSIS FOR DIFFERENT MAX_LEN VALUES:")
    print("MAX_LEN | Coverage | Truncated | Avg Padding")
    print("-" * 45)
    
    for max_len in max_len_options:
        covered = sum(1 for length in all_lengths if length <= max_len)
        coverage_pct = (covered / len(all_lengths)) * 100
        truncated = len(all_lengths) - covered
        avg_padding = max_len - np.mean([min(length, max_len) for length in all_lengths])
        
        coverage_data.append({
            'max_len': max_len,
            'coverage': coverage_pct,
            'truncated': truncated,
            'avg_padding': avg_padding
        })
        
        print(f"{max_len:7d} | {coverage_pct:7.2f}% | {truncated:9d} | {avg_padding:11.2f}")
    
    # Visualization
    plt.figure(figsize=(20, 12))
    
    # 1. Distribution histogram
    plt.subplot(2, 3, 1)
    plt.hist(all_lengths, bins=50, alpha=0.7, edgecolor='black')
    plt.axvline(stats['mean'], color='red', linestyle='--', label=f'Mean: {stats["mean"]:.1f}')
    plt.axvline(stats['median'], color='green', linestyle='--', label=f'Median: {stats["median"]:.1f}')
    plt.axvline(stats['q95'], color='orange', linestyle='--', label=f'95th percentile: {stats["q95"]:.1f}')
    plt.xlabel('Token Length')
    plt.ylabel('Frequency')
    plt.title('Distribution of Token Lengths')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 2. Box plot by dataset
    plt.subplot(2, 3, 2)
    data_for_box = [train_lengths, val_lengths, test_lengths]
    labels = ['Train', 'Validation', 'Test']
    plt.boxplot(data_for_box, labels=labels)
    plt.ylabel('Token Length')
    plt.title('Token Length Distribution by Dataset')
    plt.grid(True, alpha=0.3)
    
    # 3. Cumulative distribution
    plt.subplot(2, 3, 3)
    sorted_lengths = np.sort(all_lengths)
    percentiles = np.arange(1, len(sorted_lengths) + 1) / len(sorted_lengths) * 100
    plt.plot(sorted_lengths, percentiles)
    plt.xlabel('Token Length')
    plt.ylabel('Cumulative Percentage')
    plt.title('Cumulative Distribution of Token Lengths')
    plt.grid(True, alpha=0.3)
    
    # Add common MAX_LEN markers
    for max_len in [64, 80, 128, 256]:
        coverage = (np.searchsorted(sorted_lengths, max_len) / len(sorted_lengths)) * 100
        plt.axvline(max_len, color='red', linestyle='--', alpha=0.5)
        plt.text(max_len, coverage + 2, f'{max_len}', rotation=90, ha='center')
    
    # 4. Coverage vs MAX_LEN
    plt.subplot(2, 3, 4)
    coverages = [item['coverage'] for item in coverage_data]
    max_lens = [item['max_len'] for item in coverage_data]
    plt.plot(max_lens, coverages, marker='o', linewidth=2, markersize=8)
    plt.xlabel('MAX_LEN')
    plt.ylabel('Coverage (%)')
    plt.title('Coverage vs MAX_LEN')
    plt.grid(True, alpha=0.3)
    
    # Highlight good options
    for i, (max_len, coverage) in enumerate(zip(max_lens, coverages)):
        if coverage >= 95:
            plt.annotate(f'{max_len}\n({coverage:.1f}%)', 
                        (max_len, coverage), 
                        textcoords="offset points", 
                        xytext=(0,10), 
                        ha='center', 
                        fontsize=9,
                        bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    # 5. Efficiency analysis (Coverage vs Padding)
    plt.subplot(2, 3, 5)
    avg_paddings = [item['avg_padding'] for item in coverage_data]
    plt.scatter(avg_paddings, coverages, s=100, alpha=0.7)
    
    for i, max_len in enumerate(max_lens):
        plt.annotate(str(max_len), (avg_paddings[i], coverages[i]), 
                    textcoords="offset points", xytext=(5,5), ha='left')
    
    plt.xlabel('Average Padding')
    plt.ylabel('Coverage (%)')
    plt.title('Efficiency: Coverage vs Average Padding')
    plt.grid(True, alpha=0.3)
    
    # 6. Length distribution by class (if labels available)
    plt.subplot(2, 3, 6)
    if 'Label' in train_df.columns:
        neg_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 0]
        pos_lengths = [get_token_length(text) for text, label in 
                      zip(train_df["Text"], train_df["Label"]) if label == 1]
        
        plt.hist(neg_lengths, bins=30, alpha=0.7, label='Negative', color='red')
        plt.hist(pos_lengths, bins=30, alpha=0.7, label='Positive', color='blue')
        plt.xlabel('Token Length')
        plt.ylabel('Frequency')
        plt.title('Token Length Distribution by Class')
        plt.legend()
        plt.grid(True, alpha=0.3)
    else:
        plt.text(0.5, 0.5, 'No labels available\nfor class analysis', 
                ha='center', va='center', transform=plt.gca().transAxes,
                fontsize=12, bbox=dict(boxstyle="round", facecolor='lightgray'))
        plt.title('Class Analysis Not Available')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/max_len_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Recommendations
    print("\n RECOMMENDATIONS:")
    print("=" * 40)
    
    # Find the most efficient MAX_LEN values
    efficient_options = []
    for item in coverage_data:
        if item['coverage'] >= 90:  # At least 90% coverage
            efficiency = item['coverage'] / (item['max_len'] / stats['mean'])  # Coverage per relative length
            efficient_options.append((item['max_len'], item['coverage'], efficiency))
    
    efficient_options.sort(key=lambda x: x[2], reverse=True)  # Sort by efficiency
    
    print("Most efficient MAX_LEN options (>=90% coverage):")
    for max_len, coverage, efficiency in efficient_options[:3]:
        print(f"  MAX_LEN = {max_len}: {coverage:.1f}% coverage (efficiency: {efficiency:.2f})")
    
    # Specific recommendations based on percentiles
    print(f"\nBased on percentiles:")
    print(f"  Conservative (99% coverage): MAX_LEN = {int(stats['q99']) + 5}")
    print(f"  Balanced (95% coverage): MAX_LEN = {int(stats['q95']) + 5}")
    print(f"  Aggressive (90% coverage): MAX_LEN = {int(stats['q90']) + 5}")
    print(f"  Very aggressive (75% coverage): MAX_LEN = {int(stats['q75']) + 5}")
    
    # Return the statistics for further use
    return stats, coverage_data

print(" Starting EDA for MAX_LEN optimization...")

stats, coverage_data = analyze_text_lengths(train_df, val_df, test_df, tokenizer)


optimal_max_len = int(stats['q95']) + 5  # 95th percentile + buffer

print(f"\n SUGGESTED MAX_LEN: {optimal_max_len}")



class TweetDataset(Dataset):

    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            padding="max_length",
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long)
        }


train_dataset = TweetDataset(train_df["Text"].tolist(), train_df["Label"].tolist(), tokenizer, MAX_LEN)

val_dataset = TweetDataset(val_df["Text"].tolist(), val_df["Label"].tolist(), tokenizer, MAX_LEN)


train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)

val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)



from transformers import AutoConfig


config = AutoConfig.from_pretrained(MODEL_NAME)



config.attention_probs_dropout_prob = ATTENTION_DROPOUT  # Attention dropout

config.hidden_dropout_prob = HIDDEN_DROPOUT              # Hidden layer dropout





config.num_labels = NUM_CLASSES

config.id2label = {0: "negative", 1: "positive"}


config.label2id = {"negative": 0, "positive": 1}



model = AutoModelForSequenceClassification.from_pretrained(

    MODEL_NAME,
    config=config
).to(device)


def create_optimizer_with_weight_decay(model, learning_rate, weight_decay):

    # Separate parameters that should and shouldn't have weight decay
    decay_parameters = []
    no_decay_parameters = []
    
    for name, param in model.named_parameters():
        if not param.requires_grad:
            continue
            
        # Don't apply weight decay to bias terms and layer norm parameters
        if any(nd in name for nd in ["bias", "LayerNorm.weight", "layer_norm.weight"]):
            no_decay_parameters.append(param)
        else:
            decay_parameters.append(param)
    
    # Create parameter groups
    optimizer_grouped_parameters = [
        {
            "params": decay_parameters,
            "weight_decay": weight_decay,
            "lr": learning_rate
        },
        {
            "params": no_decay_parameters,
            "weight_decay": 0.0,
            "lr": learning_rate
        }
    ]
    

    
    return optim.AdamW(optimizer_grouped_parameters)

optimizer = create_optimizer_with_weight_decay(model, LEARNING_RATE, WEIGHT_DECAY)



total_steps = len(train_loader) * EPOCHS


scheduler = ReduceLROnPlateau(

    optimizer,
    mode='min',           # Use 'max' if you prefer to monitor val_f1
    factor=0.5,           # Reduce LR by 50%
    patience=2,           # Wait 2 epochs with no improvement before reducing LR
    threshold=1e-4,       # Minimum change to qualify as an improvement
    cooldown=1,           # Wait 1 epoch after LR is reduced before resuming
    min_lr=1e-7,          # Minimum allowed learning rate
    verbose=True          # Log changes
)



def train_epoch(model, loader, optimizer, scheduler):

    model.train()
    total_loss = 0
    batch_losses = []
    all_preds = []
    all_labels = []
    all_probs = []
    learning_rates = []  # Track learning rate changes
    
    for batch_idx, batch in enumerate(tqdm(loader, desc="Training")):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()
        batch_losses.append(loss.item())

        # Get predictions and track metrics during training
        logits = outputs.logits
        probs = F.softmax(logits, dim=1)
        preds = torch.argmax(logits, dim=1)
        
        # Collect for metrics calculation
        all_preds.extend(preds.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())
        all_probs.extend(probs[:, 1].cpu().tolist())

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        
        # Track learning rate
        current_lr = optimizer.param_groups[0]['lr']
        learning_rates.append(current_lr)
        
        optimizer.zero_grad()
        
    # Calculate metrics
    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass
    
    mcc = matthews_corrcoef(all_labels, all_preds)

    return avg_loss, batch_losses, acc, p, r, f, cm, auc_score, all_labels, all_probs, mcc, learning_rates


def evaluate(model, loader):

    model.eval()  # Important: Set to eval mode to disable dropout
    all_preds = []
    all_labels = []
    all_probs = []
    total_loss = 0
    
    with torch.no_grad():
        for batch in tqdm(loader, desc="Evaluating"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()
            
            # Get predictions and probabilities
            logits = outputs.logits
            probs = F.softmax(logits, dim=1)
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())
            all_probs.extend(probs[:, 1].cpu().tolist())

    acc = accuracy_score(all_labels, all_preds)
    p, r, f, _ = precision_recall_fscore_support(all_labels, all_preds, average="weighted")
    avg_loss = total_loss / len(loader)
    
    cm = confusion_matrix(all_labels, all_preds)
    
    # Calculate AUC if there are both positive and negative samples
    auc_score = None
    try:
        auc_score = roc_auc_score(all_labels, all_probs)
    except:
        pass

    mcc = matthews_corrcoef(all_labels, all_preds)
    
    return acc, p, r, f, avg_loss, cm, auc_score, all_labels, all_probs, mcc


def plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, 
                 model_name, batch_losses=None):
                 
    """Plot training and validation metrics"""
    plt.figure(figsize=(15, 10))
    
    # Plot losses
    plt.subplot(2, 2, 1)
    plt.plot(train_losses, label="Train Loss", marker='o')
    plt.plot(val_losses, label="Validation Loss", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid(True)
    
    # Plot accuracy
    plt.subplot(2, 2, 2)
    plt.plot(train_accs, label="Train Accuracy", marker='o')
    plt.plot(val_accs, label="Validation Accuracy", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Training and Validation Accuracy")
    plt.legend()
    plt.grid(True)
    
    # Plot F1 score
    plt.subplot(2, 2, 3)
    plt.plot(train_f1s, label="Train F1", marker='o')
    plt.plot(val_f1s, label="Validation F1", marker='x')
    plt.xlabel("Epoch")
    plt.ylabel("F1 Score")
    plt.title("Training and Validation F1 Score")
    plt.legend()
    plt.grid(True)
    
    # Plot batch-level losses if available
    if batch_losses:
        plt.subplot(2, 2, 4)
        all_batches = []
        for epoch, losses in enumerate(batch_losses):
            batches = list(range(epoch * len(losses), (epoch + 1) * len(losses)))
            all_batches.extend(batches)
            plt.plot(batches, losses, alpha=0.5, label=f"Epoch {epoch+1}")
        
        plt.xlabel("Batch")
        plt.ylabel("Loss")
        plt.title("Batch-level Loss")
        plt.legend()
        plt.grid(True)
    
    plt.tight_layout()
    plt.savefig(f"/kaggle/working/{model_name}_learning_curves.png")
    plt.show()

def plot_confusion_matrix(cm, model_name, set_name="Validation"):

    """Plot confusion matrix"""
    plt.figure(figsize=(8, 6))
    display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
    display.plot(cmap=plt.cm.Blues, values_format=".0f")
    plt.title(f"{set_name} Confusion Matrix")
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_confusion_matrix.png")
    plt.show()

def plot_roc_curve(y_true, y_probs, model_name, set_name="Validation"):

    """Plot ROC curve"""
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{set_name} ROC Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.savefig(f"/kaggle/working/{model_name}_{set_name.lower()}_roc_curve.png")
    plt.show()
    
    return roc_auc

best_f1 = 0

model_path = f"/kaggle/working/best_{MODEL_NAME.replace('-', '_')}.pth"


train_losses = []

val_losses = []

train_accs = []

val_accs = []

train_f1s = []

val_f1s = []

train_aucs = []

val_aucs = []

batch_level_losses = []


for epoch in range(EPOCHS):

    print(f"\nEpoch {epoch + 1}/{EPOCHS}")
    
    # Train and collect metrics directly during training
    train_loss, batch_losses, train_acc, train_p, train_r, train_f1, train_cm, train_auc, train_labels, train_probs, train_mcc, train_lrs = train_epoch(model, train_loader, optimizer, scheduler)
    batch_level_losses.append(batch_losses)
    
    # Evaluate only on validation set
    val_acc, val_p, val_r, val_f1, val_loss, val_cm, val_auc, val_labels, val_probs, val_mcc = evaluate(model, val_loader)

    scheduler.step(val_loss)
    
    # Store metrics
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    train_f1s.append(train_f1)
    val_f1s.append(val_f1)
    
    # Add AUC scores if available
    if train_auc is not None:
        train_aucs.append(train_auc)
    if val_auc is not None:
        val_aucs.append(val_auc)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Train  - Acc: {train_acc:.4f}, Precision: {train_p:.4f}, Recall: {train_r:.4f}, F1: {train_f1:.4f}")
    if train_auc is not None:
        print(f"         AUC: {train_auc:.4f}")
    
    print(f"Val    - Acc: {val_acc:.4f}, Precision: {val_p:.4f}, Recall: {val_r:.4f}, F1: {val_f1:.4f}, Loss: {val_loss:.4f}")
    if val_auc is not None:
        print(f"         AUC: {val_auc:.4f}")

    # Save best model
    if val_f1 > best_f1:
        best_f1 = val_f1
        torch.save(model.state_dict(), model_path)
        print("✓ Best model saved!")
        
        # Plot confusion matrix and ROC curve for best model
        plot_confusion_matrix(val_cm, MODEL_NAME)
        if val_auc is not None:
            plot_roc_curve(val_labels, val_probs, MODEL_NAME)


plot_metrics(train_losses, val_losses, train_accs, val_accs, train_f1s, val_f1s, MODEL_NAME, batch_level_losses)



model.load_state_dict(torch.load(model_path))

model.eval()


final_val_acc, final_val_p, final_val_r, final_val_f1, final_val_loss, final_val_cm, final_val_auc, final_val_labels, final_val_probs, final_val_mcc = evaluate(model, val_loader)

print("\n===== FINAL MODEL EVALUATION =====")

print(f"Val    - Acc: {final_val_acc:.4f}, Precision: {final_val_p:.4f}, Recall: {final_val_r:.4f}, F1: {final_val_f1:.4f}, MCC: {final_val_mcc:.4f}, Loss: {final_val_loss:.4f}")

if final_val_auc is not None:
    print(f"         AUC: {final_val_auc:.4f}")
    


plot_confusion_matrix(final_val_cm, MODEL_NAME, "Validation")

if final_val_auc is not None:

    plot_roc_curve(final_val_labels, final_val_probs, MODEL_NAME, "Final Validation")


class TestDataset(Dataset):

    def __init__(self, texts, tokenizer, max_len):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0)
        }

test_dataset = TestDataset(test_df["Text"].tolist(), tokenizer, MAX_LEN)

test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)


test_preds = []

with torch.no_grad():

    for batch in tqdm(test_loader, desc="Predicting"):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1)
        
        test_preds.extend(preds.cpu().tolist())


id_col = "ID" if "ID" in test_df.columns else "Id"

submission = pd.DataFrame({id_col: test_df[id_col], "Label": test_preds})

submission.to_csv(SUBMISSION_PATH, index=False)

print(f" Submission saved to: {SUBMISSION_PATH}") 
```
